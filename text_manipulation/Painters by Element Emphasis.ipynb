```python

file_path = '../data/xlsx/painters_by_element_emphasis.xlsx'
painters_by_element_emphasis_df = pd.read_excel(file_path).set_index('Element')
```

```python

def show_prompt(artist_name, verbose=False):
    prefix_str = 'kenric_tonn_'
    mask_series = painters_by_element_emphasis_df.index.map(lambda x: str(x).startswith(prefix_str))
    df = painters_by_element_emphasis_df[mask_series]
    df.index = df.index.map(lambda x: str(x).replace(prefix_str, ''))
    df = df[artist_name].to_frame()
    elements_list = []
    for i in range(-2, 3):
        mask_series = (df[artist_name] == i)
        if bool((i != 0) and df[mask_series][artist_name].shape[0]):
            if (abs(i) >= 1): prefix_str = 'emphasizing '
            if (i < 0): prefix_str = 'de-' + prefix_str
            if (abs(i) == 2): prefix_str = 'heavily ' + prefix_str
            prefix_str += nu.conjunctify_nouns(df[mask_series][artist_name].index.tolist())
            elements_list.append(prefix_str)
    print(f'Modify this picture so that it looks like a {artist_name} painting: {nu.conjunctify_nouns(elements_list)}.')
    if verbose: display(df)
```

```python

artist_name = 'Monet'
show_prompt(artist_name)
```
