```python

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

mu = 0.0
s = 1.0
y = []
x_range = np.arange(-4, 4, 0.05)
for x in x_range:
    y.append(math.exp(-(x-mu)*(x-mu)/(2*s*s))/math.sqrt(2.*math.pi*s*s))
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(111, autoscale_on=True)
path_collection = ax.scatter(x_range, y)
title = ax.annotate('$f(x \; | \; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2} } \; e^{ -\frac{(x-\mu)^2}{2\sigma^2} }$',
                   xy=(2, 0.4))
```
