```python

from IPython.display import display, Markdown
import os

file_path = '../data/html/JewishPopulation.html'
if os.path.exists(file_path):
    with open(file_path, 'r') as html_file:
        table_html = html_file.read()
        display(Markdown(table_html))
```

```python

import pandas as pd
import re
import numpy as np

jews_df = pd.read_html(file_path)[0]
numeric_columns = ['Pop1900', 'Pct1900', 'Pop1942', 'Pct1942',
                  'Pop1970', 'Pct1970', 'Pop2010', 'Pct2010']
int_columns = ['Pop1900', 'Pop1942', 'Pop1970', 'Pop2010']
float_columns = ['Pct1900', 'Pct1942', 'Pct1970', 'Pct2010']
jews_df.columns = ['Region'] + numeric_columns

def f(element):
    
    return str(element).split('[')[0]

jews_df['Region'] = jews_df['Region'].map(f)
jews_df['Pop1900'] = jews_df['Pop1900'].map(f)
num_regex = re.compile(r'[^0-9.]+')

def f(row):
    for column_name in numeric_columns:
        row[column_name] = num_regex.sub('', str(row[column_name]))
        try:
            row[column_name] = float(row[column_name])
        except Exception:
            row[column_name] = np.NaN
    
    return row
        
jews_df = jews_df.apply(f, axis=1)
jews_df
```

```python

from math import atan2, degrees
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

# Make a bunch of line plots of the various region populations
fig = plt.figure(figsize=(15, 10), dpi=80)
ax = fig.add_subplot(111, autoscale_on=True)
min_pop = 800000
mask_series = (jews_df['Pop1900'] > min_pop) | (jews_df['Pop2010'] > min_pop)
for _, row_series in jews_df[mask_series].iterrows():
    dots = ax.plot([float(i[3:]) for i in int_columns], row_series[int_columns], label=row_series['Region'])

lines = plt.gca().get_lines()
kwargs = {}
kwargs['zorder'] = 2.5

ax = lines[0].axes
labLines = []
labels = []

#Take only the lines which have labels other than the default ones
for line in lines:
    label = line.get_label()
    if "_line" not in label:
        labLines.append(line)
        labels.append(label)

xmin, xmax = ax.get_xlim()
xvals = np.linspace(xmin, xmax, len(labLines)+2)[1:-1]

for line, x, label in zip(labLines, xvals, labels):
    ax = line.axes
    xdata = line.get_xdata()
    ydata = line.get_ydata()
    
    # Handle cases where the line doesn't begin on the left
    xdata_min = xdata[list(ydata).index(next(y for y in ydata if y==y))]
    if xdata[0] < xdata_min:
        x = np.linspace(xdata_min, xmax, 3)[1]

    if (x < xdata[0]) or (x > xdata[-1]):
        print('x label location is outside data range!')

    # Find corresponding y co-ordinate and angle of the line
    ip = 1
    for i in range(len(xdata)):
        if x < xdata[i]:
            ip = i
            break

    y = ydata[ip-1] + (ydata[ip]-ydata[ip-1])*(x-xdata[ip-1])/(xdata[ip]-xdata[ip-1])

    if not label:
        label = line.get_label()

    #Compute the slope
    dx = xdata[ip] - xdata[ip-1]
    dy = ydata[ip] - ydata[ip-1]
    ang = degrees(atan2(dy, dx))

    #Transform to screen co-ordinates
    pt = np.array([x, y]).reshape((1, 2))
    trans_angle = ax.transData.transform_angles(np.array((ang, )), pt)[0]

    #Set a bunch of keyword arguments
    kwargs['color'] = line.get_color()

    if ('horizontalalignment' not in kwargs) and ('ha' not in kwargs):
        kwargs['ha'] = 'center'

    if ('verticalalignment' not in kwargs) and ('va' not in kwargs):
        kwargs['va'] = 'center'

    if 'backgroundcolor' not in kwargs:
        kwargs['backgroundcolor'] = ax.get_facecolor()

    if 'clip_on' not in kwargs:
        kwargs['clip_on'] = True
    
    inline_label = ax.text(x, y, label, rotation=trans_angle, **kwargs)
#legend = ax.legend(loc='center right')
```
