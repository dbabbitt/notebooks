
## Load the data

```python

import re

lf_regex = re.compile(r'\n')
st_regex = re.compile(r'\bSaint\b')
rep_regex = re.compile(r'\bRepublic\b')
is_regex = re.compile(r'\bIsl(e|ands?)\b')
lao_regex = re.compile(r'\bLao\b')
```

```python

import os.path
import pandas as pd

income_file = './data/csv/income_df.csv'
if os.path.isfile(income_file):
    encoding = ['latin1', 'iso8859-1', 'utf-8'][1]
    income_df = pd.read_csv(income_file, encoding=encoding)
    income_df.set_index(keys='Country', inplace=True)
else:
    income_df = pd.read_csv('./data/csv/indicator gapminder gdp_per_capita_ppp.csv')
    income_df = income_df.dropna(how='all').T.dropna(how='all').T.dropna(subset=['GDP per capita'])

    income_df['GDP per capita'] = income_df['GDP per capita'].map(lambda x: lf_regex.sub(r'', x))

    income_df['GDP per capita'] = income_df['GDP per capita'].map(lambda x: st_regex.sub(r'St.', x))

    income_df['GDP per capita'] = income_df['GDP per capita'].map(lambda x: rep_regex.sub(r'Rep.', x))

    income_df['GDP per capita'] = income_df['GDP per capita'].map(lambda x: is_regex.sub(r'Is.', x))

    income_df['GDP per capita'] = income_df['GDP per capita'].map(lambda x: lao_regex.sub(r'Laos', x))

    income_df.set_index(keys='GDP per capita', inplace=True)

income_df.fillna(0, inplace=True)
income_df.index.name = 'Country'
income_df.to_csv(income_file, index=True)
income_df.sample(5).T.head(10).T
```

```python

fertility_file = './data/csv/fertility_df.csv'
if os.path.isfile(fertility_file):
    encoding = ['latin1', 'iso8859-1', 'utf-8'][1]
    fertility_df = pd.read_csv(fertility_file, encoding=encoding)
    fertility_df.set_index(keys='Country', inplace=True)
else:
    fertility_df = pd.read_csv('./data/csv/indicator undata total_fertility.csv')
    fertility_df = fertility_df.dropna(how='all').T.dropna(how='all').T.dropna(subset=['Total fertility rate'])

    fertility_df['Total fertility rate'] = fertility_df['Total fertility rate'].map(lambda x: lf_regex.sub(r'', x))

    fertility_df['Total fertility rate'] = fertility_df['Total fertility rate'].map(lambda x: st_regex.sub(r'St.', x))

    fertility_df['Total fertility rate'] = fertility_df['Total fertility rate'].map(lambda x: rep_regex.sub(r'Rep.', x))

    fertility_df['Total fertility rate'] = fertility_df['Total fertility rate'].map(lambda x: is_regex.sub(r'Is.', x))

    fertility_df['Total fertility rate'] = fertility_df['Total fertility rate'].map(lambda x: lao_regex.sub(r'Laos', x))

    fertility_df.set_index(keys='Total fertility rate', inplace=True)

fertility_df.fillna(0, inplace=True)
fertility_df.index.name = 'Country'
fertility_df.to_csv(fertility_file, index=True)
fertility_df.sample(5).T.sample(15).T
```

```python

population_file = './data/csv/population_df.csv'
if os.path.isfile(population_file):
    encoding = ['latin1', 'iso8859-1', 'utf-8'][1]
    population_df = pd.read_csv(population_file, encoding=encoding)
    population_df.set_index(keys='Country', inplace=True)
    population_df.fillna(0, inplace=True)
else:
    encoding = ['latin1', 'iso8859-1', 'utf-8'][2]
    population_df = pd.read_csv('./data/csv/indicator gapminder population.csv', error_bad_lines=False, encoding=encoding)
    population_df = population_df.dropna(how='all').T.dropna(how='all').T.dropna(subset=['Total population'])

    population_df['Total population'] = population_df['Total population'].map(lambda x: lf_regex.sub(r'', x))

    population_df['Total population'] = population_df['Total population'].map(lambda x: st_regex.sub(r'St.', x))

    population_df['Total population'] = population_df['Total population'].map(lambda x: rep_regex.sub(r'Rep.', x))

    population_df['Total population'] = population_df['Total population'].map(lambda x: is_regex.sub(r'Is.', x))

    population_df['Total population'] = population_df['Total population'].map(lambda x: lao_regex.sub(r'Laos', x))

    population_df.set_index(keys='Total population', inplace=True)
    population_df.fillna(0, inplace=True)
    int_regex = re.compile(r'[^0-9.]+')
    population_df = population_df.applymap(lambda x: int(int_regex.sub(r'', str(x))))
    
population_df.index.name = 'Country'
population_df.to_csv(population_file, index=True)
population_df.sample(5).T.sample(10).T
```

```python

from bokeh.palettes import Spectral6

regions_file = './data/csv/regions_df.csv'
if os.path.isfile(population_file):
    encoding = ['latin1', 'iso8859-1', 'utf-8'][1]
    regions_df = pd.read_csv(regions_file, encoding=encoding)
    regions_df.set_index(keys='Country', inplace=True)
else:
    from bokeh.sampledata.gapminder import regions as regions_df
    
    regions_df['Country'] = regions_df['Country'].map(lambda x: st_regex.sub(r'St.', x))

    regions_df['Country'] = regions_df['Country'].map(lambda x: rep_regex.sub(r'Rep.', x))

    regions_df['Country'] = regions_df['Country'].map(lambda x: is_regex.sub(r'Is.', x))

    regions_df['Country'] = regions_df['Country'].map(lambda x: lao_regex.sub(r'Laos', x))

regions_df['Group'] = regions_df['Group'].astype('category')
regions_list = list(regions_df['Group'].cat.categories)
regions_df.index.name = 'Country'

def get_color(r):
    return Spectral6[regions_list.index(r['Group'])]

regions_df['region_color'] = regions_df.apply(get_color, axis=1)

print(regions_list)
regions_df.to_csv(regions_file, index=True)
regions_df.sample(5)
```

```python

from bokeh.palettes import Spectral7
#from sympy.ntheory import factorint

religions_file = './data/csv/religions_df.csv'
if os.path.isfile(religions_file):
    encoding = ['latin1', 'iso8859-1', 'utf-8'][1]
    religions_df = pd.read_csv(religions_file, encoding=encoding)
    religions_df.set_index(keys='Country', inplace=True)
else:
    religions_file = './data/csv/religions.csv'
    encoding = ['latin1', 'iso8859-1', 'utf-8'][2]
    religions_df = pd.read_csv(religions_file, encoding=encoding)
    religions_df = religions_df.applymap(lambda x: str(x).strip())

    pop_col_list = ['Population', 'Christian', 'Muslim', 'Unaffiliated', 'Hindu', 'Buddhist', 'FolkReligion', 'OtherReligion', 
                    'Jewish']
    split_regex = re.compile(r'[^0-9.]+')
    religions_df[pop_col_list] = religions_df[pop_col_list].applymap(lambda x: int(split_regex.sub(r'', x)))

    perc_column_list = [x for x in religions_df.columns if re.search(r'%', x)]
    religions_df[perc_column_list] = religions_df[perc_column_list].applymap(lambda x: float(split_regex.split(str(x))[0]))

    jewishness_series = sorted(religions_df['Jewish%'].unique())
    for i in range(0, len(jewishness_series), 4):
        match_series = (religions_df['Jewish%'].isin(jewishness_series[i:i+4]))
        religions_df.loc[match_series, 'Jewishness'] = i

    religions_df['Jewishness'] = religions_df['Jewishness'].astype('category')
    religions_list = list(religions_df['Jewishness'].cat.categories)

    religions_df['Country'] = religions_df['Country'].map(lambda x: lf_regex.sub(r'', x))

    religions_df['Country'] = religions_df['Country'].map(lambda x: st_regex.sub(r'St.', x))

    religions_df['Country'] = religions_df['Country'].map(lambda x: rep_regex.sub(r'Rep.', x))

    religions_df['Country'] = religions_df['Country'].map(lambda x: is_regex.sub(r'Is.', x))

    religions_df['Country'] = religions_df['Country'].map(lambda x: lao_regex.sub(r'Laos', x))

    religions_df.set_index(keys='Country', inplace=True)
    
religions_df.index.name = 'Country'

def get_color(r):
    return Spectral7[religions_list.index(r['Jewishness'])]

religions_df['jewishness_color'] = religions_df.apply(get_color, axis=1)
religions_df.to_csv(religions_file, index=True)

print(religions_list)
religions_df.sample(5).T
```

```python

from difflib import SequenceMatcher
import time


def similar(a, b):
    return SequenceMatcher(None, str(a), str(b)).ratio()

#Check the closest names for typos
def check_for_typos(first_list, second_list):
    rows_list = []
    for first_item in first_list:
        max_similarity = 0.0
        max_item = first_item
        for second_item in second_list:
            this_similarity = similar(first_item, second_item)
            if this_similarity > max_similarity:
                max_similarity = this_similarity
                max_item = second_item

        # Get input row in dictionary format; key = col_name
        row_dict = {}
        row_dict['first_item'] = first_item
        row_dict['second_item'] = max_item
        row_dict['max_similarity'] = max_similarity

        rows_list.append(row_dict)

    column_list = ['first_item', 'second_item', 'max_similarity']
    name_similarities_df = pd.DataFrame(rows_list, columns=column_list)
    
    return name_similarities_df
```


So, what I can do here is eyeball the list and find a `max_similarity` low enough to include only true comparisons above it.

```python

# Closest fertility countries to the countries in the income dataframe that are not paired up
t0 = time.time()
name_similarities_df = check_for_typos(income_df.index.difference(fertility_df.index), fertility_df.index)
t1 = time.time()
print(t1-t0, time.ctime(t1))
name_similarities_df.sort_values(['max_similarity'], ascending=False).head(10)
```


## Finish prepping data

```python

# Get common countries across all data: fertility is the lowest comon denominator
income_df = income_df.drop(income_df.index.difference(fertility_df.index))
population_df = population_df.drop(population_df.index.difference(fertility_df.index))
regions_df = regions_df.drop(regions_df.index.difference(fertility_df.index))
religions_df = religions_df.drop(religions_df.index.difference(fertility_df.index))
```

```python

# Make the column names ints not strings for handling
columns_list = list(fertility_df.columns)
years_list = list(range(int(columns_list[0]), int(columns_list[-1])))
rename_dict = dict(zip(columns_list, years_list))

fertility_df = fertility_df.rename(columns=rename_dict)
fertility_df.to_csv('./data/csv/fertility_df.csv', index=True)
income_df = income_df.rename(columns=rename_dict)
income_df.to_csv('./data/csv/income_df.csv', index=True)
population_df = population_df.rename(columns=rename_dict)
population_df.to_csv('./data/csv/population_df.csv', index=True)
regions_df = regions_df.rename(columns=rename_dict)
regions_df.to_csv('./data/csv/regions_df.csv', index=True)
religions_df = religions_df.rename(columns=rename_dict)
religions_df.to_csv('./data/csv/religions_df.csv', index=True)
```

```python

import numpy as np

# Turn population into bubble sizes. Use min_size and factor to tweak
scale_factor = 400
population_size_df = np.sqrt(population_df / np.pi) / scale_factor
min_size = 3
population_size_df = population_size_df.where(population_size_df >= min_size).fillna(min_size)

#region_color_series = regions_df['region_color']
region_color_series = religions_df['jewishness_color']
region_color_series.name = 'region_color'
```

```python

# Change zeroes to nulls
def f(x):
    if x == 0:
        x = np.nan
    
    return x

income_df = income_df.applymap(f)
fertility_df = fertility_df.applymap(f)
population_df = population_df.applymap(f)
```

```python

from bokeh.models import Range1d
from bokeh.models import DataRange1d
import math

# Children per woman (total fertility)
x_low = int(math.floor(fertility_df.min().min()))
x_high = int(math.ceil(fertility_df.max().max()))
xdr = Range1d(x_low-0.5*x_low, x_high+0.1*x_high)
x_interval = int(math.ceil((x_high - x_low) / 9))

# Personal income (GDP per capita)
y_low = int(math.floor(income_df.min().min()))
y_high = int(math.ceil(income_df.max().max()))
ydr = DataRange1d(y_low-0.5*y_low, y_high+0.5*y_high)
y_interval = int(math.floor((y_high - y_low) / 5))

AXIS_FORMATS = dict(
    minor_tick_in=None,
    minor_tick_out=None,
    major_tick_in=None,
    major_label_text_font_size="10pt",
    major_label_text_font_style="normal",
    axis_label_text_font_size="10pt",

    axis_line_color='#AAAAAA',
    major_tick_line_color='#AAAAAA',
    major_label_text_color='#666666',

    major_tick_line_cap="round",
    axis_line_cap="round",
    axis_line_width=1,
    major_tick_line_width=1,
)
```

```python

import numpy as np

fertility_series = fertility_df['2015']
fertility_series.name = 'fertility'

income_series = income_df['2015']
income_series.name = 'income'

population_series = population_size_df['2015']
population_series.name = 'population'

new_df = pd.concat([fertility_series, income_series, population_series, region_color_series], axis=1)

new_df.head()
```

```python

from bokeh.models import Plot
from bokeh.models import LogScale

# Build the plot
plot = Plot(
    
    # Children per woman (total fertility)
    x_range=xdr,
    
    # Personal income (GDP per capita)
    y_range=ydr,
    y_scale=LogScale(),
    
    plot_width=800,
    plot_height=400,
    outline_line_color=None,
    toolbar_location=None, 
    min_border=20,
)
```

```python

from bokeh.models import LinearAxis
from bokeh.models import SingleIntervalTicker
from bokeh.models import LogAxis
from bokeh.models import LogTicker

# Build the axes
xaxis = LinearAxis(ticker=SingleIntervalTicker(interval=x_interval), axis_label="Children per woman (total fertility)", 
                   **AXIS_FORMATS)
yaxis = LogAxis(ticker=LogTicker(mantissas=[1, 5]), axis_label="Personal income (GDP per capita)", 
                   **AXIS_FORMATS)

plot.add_layout(xaxis, 'below')
plot.add_layout(yaxis, 'left')
```

```python

from bokeh.models import Plot
from bokeh.models import LogScale

from bokeh.models import LinearAxis
from bokeh.models import SingleIntervalTicker
from bokeh.models import LogAxis
from bokeh.models import LogTicker

from bokeh.models import Text
from bokeh.models import ColumnDataSource

from bokeh.models import Circle

from bokeh.io import export_png

import os
import imageio

import time

t0 = time.time()
png_dir = "./saves/png/"
for year in years_list:
    
    if (year in fertility_df) and (year in income_df) and (year in population_size_df):

        fertility_series = fertility_df[year]
        fertility_series.name = 'fertility'

        income_series = income_df[year]
        income_series.name = 'income'

        population_series = population_size_df[year]
        population_series.name = 'population'

        new_df = pd.concat([fertility_series, income_series, population_series, region_color_series], axis=1)

        # Build the plot
        plot = Plot(

            # Children per woman (total fertility)
            x_range=xdr,

            # Personal income (GDP per capita)
            y_range=ydr,
            y_scale=LogScale(),
            
            plot_width=800,
            plot_height=400,
            outline_line_color=None,
            toolbar_location=None, 
            min_border=20,
        )

        # Build the axes
        xaxis = LinearAxis(ticker=SingleIntervalTicker(interval=x_interval), axis_label="Children per woman (total fertility)", 
                           **AXIS_FORMATS)
        yaxis = LogAxis(ticker=LogTicker(), axis_label="Personal income (GDP per capita)", 
                           **AXIS_FORMATS)
        plot.add_layout(xaxis, 'below')
        plot.add_layout(yaxis, 'left')

        # Add the background year text
        text_source = ColumnDataSource({'year': ['%s' % year]})
        text = Text(x=1, y=150, text='year', text_font_size='150pt', text_color='#EEEEEE')
        plot.add_glyph(text_source, text)

        # Add the circle
        renderer_source = ColumnDataSource(new_df)
        circle_glyph = Circle(
            x='fertility', y='income', size='population',
            fill_color='region_color', fill_alpha=0.8, 
            line_color='#7c7e71', line_width=0.5, line_alpha=0.5)
        circle_renderer = plot.add_glyph(renderer_source, circle_glyph)

        # Add the legend
        text_x = 7
        for i, region in enumerate(regions_list):
            plot.add_glyph(Text(x=text_x, y=int(10**(-0.15*i+5)), text=[region], text_font_size='10pt', text_color='#666666'))
            plot.add_glyph(Circle(x=text_x - 0.1, y=int(10**(-0.15*i+5.08)), fill_color=Spectral6[i], size=10, line_color=None, 
                                  fill_alpha=0.8))

        # Save as PNG
        export_png(plot, filename=png_dir+"plot_%s.png" % year)
        
images = []
for subdir, dirs, files in os.walk(png_dir):
    for file in files:
        filepath = os.path.join(subdir, file)
        if filepath.endswith(".png"):
            images.append(imageio.imread(filepath))
imageio.mimsave('./saves/gif/movie.gif', images)

t1 = time.time()
print(t1-t0, time.ctime(t1))
```
