```python

import requests
import os
from bs4 import BeautifulSoup
import re

from selenium import webdriver
from base64 import b64encode
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import selenium.webdriver.support.ui as ui
```

```python

siteUrl = "https://www.nsa.gov/careers/"
print(siteUrl)
```

```python

# Retrieve the page with tag results and set it up to be scraped
fp = webdriver.FirefoxProfile()
driver = webdriver.Firefox(fp)
driver.get(siteUrl)

wait = ui.WebDriverWait(driver, 10)

page_html = driver.page_source
driver.quit()
```

```python

for parser in ['lxml', 'html5lib', 'html.parser']:
    soup = BeautifulSoup(page_html, parser)
    result = soup.find('body').text
    print(parser, len(result))
```

```python

soup = BeautifulSoup(page_html, 'html5lib')
result = soup.find('a')
print(result)
```

```python

soup = BeautifulSoup(page_html, 'lxml')
for anchor in soup.find_all("a"):
    if "apply" in anchor.text:
        print(anchor)
        driver2 = webdriver.Firefox(fp)
        driver2.get(siteUrl)
        wait2 = ui.WebDriverWait(driver, 10)
        page_html2 = driver2.page_source
        driver2.quit()
```

```python

soup = BeautifulSoup(page_html2, 'lxml')
result = soup.find('body')
print(result.text)
```
