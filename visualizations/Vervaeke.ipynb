
### Create the layout dictionary

```python

# You have to rerun this if you added nodes or edges
li = nx.fruchterman_reingold_layout(vervaeke_digraph).items()
print([f'nx.{fn}' for fn in dir(nx) if fn.endswith('_layout')])
```

```python

# You have to rerun this if you added nodes
rows_list = [{'node_name': node_name, 'layout_x': pos_array[0], 'layout_y': pos_array[1]} for node_name, pos_array in li]
df = DataFrame(rows_list).sort_values('layout_x')
df['x_tick'] = [int(round(el)) for el in pd.cut(np.array([left_lim, right_lim]), len(vervaeke_digraph.nodes)+1, retbins=True)[1]][1:-1]
df = df.sort_values('layout_y')
df['y_tick'] = [int(round(el)) for el in pd.cut(np.array([bottom_lim, top_lim]), len(vervaeke_digraph.nodes)+1, retbins=True)[1]][1:-1]
```

```python

# You have to rerun this if you added nodes or edges
vervaeke_layout_dict = {}
for row_index, row_series in df.iterrows():
    node_name = row_series.node_name
    layout_x = row_series.x_tick
    layout_y = row_series.y_tick
    vervaeke_layout_dict[node_name] = np.array([float(layout_x), float(layout_y)])
```


### Draw the graph using the layout

```python

# Draw the graph using the layout
vervaeke_fig = plt.figure(figsize=(fig_width, fig_width*aspect_ratio))
ax = vervaeke_fig.add_subplot(111)
plt.axis('off')
plt.xticks([], [])
plt.yticks([], [])
vervaeke_fig.suptitle('Rationality', fontsize=24)

# Make fat arrows
arrowstyle = mpatches.ArrowStyle.Simple(head_length=2., head_width=1.5, tail_width=0.05)

# Make the nodes blue
node_collection = nx.draw_networkx_nodes(G=vervaeke_digraph, pos=vervaeke_layout_dict, alpha=0.25,
                                         node_color='b', node_size=150)

edge_collection = nx.draw_networkx_edges(G=vervaeke_digraph, pos=vervaeke_layout_dict, alpha=0.25,
                                         arrowstyle=arrowstyle)
labels_collection = nx.draw_networkx_labels(G=vervaeke_digraph, pos=vervaeke_layout_dict, font_size=10)
axes_subplot = vervaeke_fig.get_axes()[0]
# xlim_tuple = axes_subplot.set_xlim(left=-1500, right=1500)
```
