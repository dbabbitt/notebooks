
Each nucleotide (for RNA) or base pair (for DNA) represents 2 bits of information because there are four possible bases (A, T, C, G for DNA or A, U, C, G for RNA). Since $\log_2(4) = 2$, each base encodes 2 bits of information.

```python

import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['RNA viruses', 'DNA viruses', 'Parasitic bacteria', 'Parasitic fungi', 'Protozoa', 'Helminths', 'Arthropods']
lower_ends = [7000, 10000, 1000000, 20000000, 20000000, 200000000, 200000000]
upper_ends = [64000, 460000, 4000000, 200000000, 120000000, 2000000000, 4000000000]

# Create the plot
fig, ax = plt.subplots(figsize=(12, 6))

# Create the bar chart
x = np.arange(len(categories))
for i, (low, high, label) in enumerate(zip(lower_ends, upper_ends, categories)):
    ax.bar(
        x[i], high - low, bottom=low, width=0.5,
        edgecolor='black', linewidth=1.5, label=label
    )

# Customize the plot
ax.set_ylabel('Bits of information encoded in the base pairs (log scale)')
ax.set_title('Genome Sizes of Various Organisms')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

# Set y-axis to logarithmic scale
ax.set_yscale('log')

# Add gridlines
ax.grid(True, which="both", ls="-", alpha=0.2)

# Adjust layout and display
plt.tight_layout()
plt.show()
```
