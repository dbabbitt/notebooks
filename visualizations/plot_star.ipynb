
# Star

Create and draw a star with varying node properties.


```python

from pygraphviz import AGraph

A = AGraph()

# Set some default node attributes
A.node_attr['style'] = 'filled'
A.node_attr['shape'] = 'circle'
A.node_attr['fixedsize'] = 'true'
A.node_attr['fontcolor'] = '#FFFFFF'
```

```python

# Make a star in shades of red
for i in range(1, 16):
    A.add_edge(0, i)
    n = A.get_node(i)
    n.attr['fillcolor'] = '#%2x0000' % (i * 16)
    n.attr['height'] = '%s' % (i / 16.0 + 0.5)
    n.attr['width'] = '%s' % (i / 16.0 + 0.5)
```

```python

print(A.string())  # print to screen
```

```python

import os

dot_dir = '../saves/dot'
os.makedirs(name=dot_dir, exist_ok=True)
file_path = os.path.join(dot_dir, 'star.dot')
A.write(file_path)  # write to simple.dot
```

```python

png_dir = '../saves/png'
file_path = os.path.join(png_dir, 'star.png')
A.draw(file_path, prog='circo')  # draw to png using circo layout
```
