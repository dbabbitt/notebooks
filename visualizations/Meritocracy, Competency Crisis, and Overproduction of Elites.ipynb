```python

from markdownify import markdownify as md
import pyperclip

html = '''<div class="tdb-block-inner td-fix-index">
<p>The books of the Bible were written over a period of over a millennium and a half, with the oldest parts composed around 1500-1400 BC and the most recent books from the late first century AD. These books cover the entire span of time from the creation of the universe to its final consummation â€” but most focus on roughly the period of their composition.&nbsp;</p>



<p>This article will provide the best approximation for when each book of the Bible was written, based on both traditional accounts and the latest scholarship.&nbsp;</p><div class="td-a-ad id_inline_ad0 id_ad_content-horiz-center"><span class="td-adspot-title"></span><div id="FreeStarVideoAdContainer">
 <div id="freestar-video-parent">
  <div id="freestar-video-child"><div style="width: 100%; height: 100%;"><div id="primisPlayerContainerDiv" class="primisslate"><iframe marginwidth="0" marginheight="0" frameborder="0" scrolling="no" title="Primis Frame" id="google_ads_iframe_dummy_sekindoParent694" style="width: 1px; height: 1px; position: absolute;"></iframe><div id="primis_container_div"><div id="placeHolder" style="position: relative; overflow: hidden; width: 400px; height: 225px;"><div style="position: absolute; top: 50%; left: 50%; transform: translate(-50%, -50%) scale(1.60714); display: block;"><div style="width: 100px;height: 100px;position: absolute;top: 50%;left: 50%;transform: translate(-50%, -50%);border: 5px solid #eeeeee;border-radius: 500px;"></div><div style="position: absolute;top: 50%;left: 4px;transform: translate(-50%, -50%);width: 0;height: 0;border-top: 15px solid transparent;border-left: 27px solid #eeeeee;border-bottom: 15px solid transparent;"></div></div></div><div id="primis_playerSekindoSPlayer6712b8054999e" style="overflow: visible; position: fixed; width: 400px; height: 225px; box-shadow: rgba(0, 0, 0, 0.4) 1px 1px 6px 3px; border-radius: 1px; display: block; z-index: 2147483646; right: 10px; bottom: 105px;"><div id="Player-Div-SekindoSPlayer6712b8054999e" style="width: max-content; height: max-content; top: auto; left: auto; position: static; transform: none;"><div id="closeContainer" style="display: block; height: 21px; width: 100%; position: absolute; pointer-events: none; z-index: 5; top: -21px;"><div id="closeIcon" style="height: 16px; width: 16px; position: relative; float: right;"><div style="pointer-events: auto; width: 60px;height: 60px;position: absolute;top: 50%;left: 50%;transform: translate(-50%, -50%);z-index: -1;"></div><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width:100%; height:100%; cursor: pointer" viewBox="0 0 768 768" enable-background="new 0 0 768 768" x="0px" y="0px" version="1.1" id="Layer_1" xmlns:xlink="http://www.w3.org/1999/xlink" xml:space="preserve"><circle fill="#ffffff" opacity="0.25" cx="384" cy="384" r="341"></circle><g><path fill="#000000" d="M655.5,112.5C582.9,39.9,486.6,0,384,0S185.1,39.9,112.5,112.5S0,281.4,0,384s39.9,198.9,112.5,271.5S281.4,768,384,768 s198.9-39.9,271.5-112.5S768,486.6,768,384S728.1,185.1,655.5,112.5z M624.6,624.6C560.3,689,474.8,724.4,384,724.4 S207.7,689,143.4,624.6C10.7,492,10.7,276,143.4,143.4C207.7,79,293.2,43.6,384,43.6S560.3,79,624.6,143.4 C757.3,276,757.3,492,624.6,624.6z" class="active-path" data-old_color="#A96D6D" data-original="#000000"></path><path fill="#000000" d="M553.2,214.8c-8.6-8.6-22.3-8.6-30.9,0L384,353.1L245.7,214.8c-8.6-8.6-22.3-8.6-30.9,0c-8.6,8.6-8.6,22.3,0,30.9 L353.1,384L214.8,522.3c-8.6,8.6-8.6,22.3,0,30.9c4.2,4.2,9.9,6.5,15.4,6.5s11.2-2.1,15.4-6.5l138.3-138.3l138.3,138.3 c4.2,4.2,9.9,6.5,15.4,6.5c5.7,0,11.2-2.1,15.4-6.5c8.6-8.6,8.6-22.3,0-30.9L414.9,384l138.3-138.3 C561.8,237.1,561.8,223.4,553.2,214.8z" class="active-path" data-old_color="#A96D6D" data-original="#000000"></path></g></svg><div id="closeIconHit" style="height: 16px; width: 16px; pointer-events: auto; position: absolute; top: 0px; left: 0px; cursor: pointer;"></div></div></div><div id="Video-Div-SekindoSPlayer6712b8054999e" style="position: relative;"><div id="Video-iFrame-SekindoSPlayer6712b8054999e" style="overflow: hidden; position: relative; width: 400px; height: 225px;"><div id="videoContainerDiv" style="width: 100%; height: 100%; background: black; position: relative; z-index: 1;"><div style="width: 100%; height: 100%; display: block;"><video playsinline="" crossorigin="anonymous" style="width: 100%; height: 100%;" src="blob:https://www.biblegateway.com/04160ed6-e105-4b86-ac5c-3ffadebdccf4"></video></div><div style="position: absolute; top: 50%; left: 50%; transform: translate(-50%, -50%) scale(1.14286, 0.5625); z-index: 3; visibility: visible; width: 350px; height: 400px; display: none;"><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px 0px; position: absolute; left: 0px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -50px; position: absolute; left: 0px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -100px; position: absolute; left: 0px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -150px; position: absolute; left: 0px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -200px; position: absolute; left: 0px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -250px; position: absolute; left: 0px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -300px; position: absolute; left: 0px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: 0px -350px; position: absolute; left: 0px; top: 350px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px 0px; position: absolute; left: 50px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -50px; position: absolute; left: 50px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -100px; position: absolute; left: 50px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -150px; position: absolute; left: 50px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -200px; position: absolute; left: 50px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -250px; position: absolute; left: 50px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -300px; position: absolute; left: 50px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -50px -350px; position: absolute; left: 50px; top: 350px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px 0px; position: absolute; left: 100px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -50px; position: absolute; left: 100px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -100px; position: absolute; left: 100px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -150px; position: absolute; left: 100px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -200px; position: absolute; left: 100px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -250px; position: absolute; left: 100px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -300px; position: absolute; left: 100px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -100px -350px; position: absolute; left: 100px; top: 350px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px 0px; position: absolute; left: 150px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -50px; position: absolute; left: 150px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -100px; position: absolute; left: 150px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -150px; position: absolute; left: 150px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -200px; position: absolute; left: 150px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -250px; position: absolute; left: 150px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -300px; position: absolute; left: 150px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -150px -350px; position: absolute; left: 150px; top: 350px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px 0px; position: absolute; left: 200px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -50px; position: absolute; left: 200px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -100px; position: absolute; left: 200px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -150px; position: absolute; left: 200px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -200px; position: absolute; left: 200px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -250px; position: absolute; left: 200px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -300px; position: absolute; left: 200px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -200px -350px; position: absolute; left: 200px; top: 350px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px 0px; position: absolute; left: 250px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -50px; position: absolute; left: 250px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -100px; position: absolute; left: 250px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -150px; position: absolute; left: 250px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -200px; position: absolute; left: 250px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -250px; position: absolute; left: 250px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -300px; position: absolute; left: 250px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -250px -350px; position: absolute; left: 250px; top: 350px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px 0px; position: absolute; left: 300px; top: 0px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -50px; position: absolute; left: 300px; top: 50px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -100px; position: absolute; left: 300px; top: 100px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -150px; position: absolute; left: 300px; top: 150px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -200px; position: absolute; left: 300px; top: 200px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -250px; position: absolute; left: 300px; top: 250px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -300px; position: absolute; left: 300px; top: 300px;"></div><div style="width: 51px; height: 51px; background-image: url(&quot;https://video.primis.tech/uploads/cn1/video/users/converted/27975/video_666a9f58cd6b4362635916/vid666aad0cdf20f457233928.jpg?cbuster=1718267151&quot;); background-repeat: no-repeat; background-size: 350px 400px; background-position: -300px -350px; position: absolute; left: 300px; top: 350px;"></div></div></div><div id="adContainerDiv" style="width: auto; height: auto; position: absolute; margin: auto; display: block; z-index: 2; top: 0px; left: 0px; cursor: pointer; background: black;"><div id="adVpaid" style="width: 100%; height: 100%; position: absolute; margin: auto; display: block; top: 0px; left: 0px; cursor: pointer; background: black;"></div><div id="adIma" style="width: 100%; height: 100%; position: absolute; margin: auto; display: block; top: 0px; left: 0px; cursor: pointer; background: black;"><div id="imaSlotContainer" style="width: 400px; height: 225px; margin-left: auto; margin-right: auto; z-index: -3; position: absolute; transform: translate(-3000px); overflow: hidden; background: black;"><div style="position: absolute;"><div style="display: none;"><video title="Advertisement" webkit-playsinline="true" playsinline="true" style="background-color: rgb(0, 0, 0); position: absolute; width: 100%; height: 100%; left: 0px; top: 0px;"></video><div style="position: absolute; width: 100%; height: 100%; left: 0px; top: 0px;"></div></div><div style="display: none;"><video title="Advertisement" webkit-playsinline="true" playsinline="true" style="background-color: rgb(0, 0, 0); position: absolute; width: 100%; height: 100%; left: 0px; top: 0px;"></video><div style="position: absolute; width: 100%; height: 100%; left: 0px; top: 0px;"></div></div><iframe src="https://imasdk.googleapis.com/js/core/bridge3.673.0_en.html#goog_2017133358" allowfullscreen="" allow="autoplay;attribution-reporting" id="goog_2017133358" title="Advertisement" style="border: 0px; opacity: 0; margin: 0px; padding: 0px; position: relative; color-scheme: light;"></iframe><iframe title="Advertisement" style="display: none;"></iframe></div><video width="0" height="0" id="adVideoElement" playsinline="" autoplay="" style="top: 0px; z-index: -1; position: absolute; left: 0px; background-color: black;"></video></div></div></div><div id="layoutContainerDiv" style="width: 100%; height: 100%; position: absolute; display: inline-block; z-index: 3; top: 0px; left: 0px; pointer-events: none;"><div id="layoutDesign" style="opacity: 1; overflow: hidden; position: absolute; left: 0px; top: 0px; width: 100%; height: 100%; z-index: 111; pointer-events: none;"><div id="adCover" style="position: absolute; top: 0px; left: 0px; z-index: 100; cursor: pointer; width: 100%; height: 100%; display: none; visibility: visible; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="transparentCover" style="cursor: pointer; width: 100%; height: 100%; visibility: visible; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="transparentInner" style="cursor: pointer; visibility: inherit; width: 100%; height: 100%; background-color: rgba(0, 0, 0, 0.4); opacity: 0; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div></div><div id="contextMenuPrimis" style="z-index: 100000; position: absolute; width: 132px; height: 25px; cursor: pointer; background-color: rgba(0, 0, 0, 0.7); border-radius: 6px; border-width: 1px; border-style: solid; border-color: white; display: none; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="contextMenuPoweredBy" style="position: relative; float: left; line-height: 25px; margin-left: 5px; font-family: Roboto; font-size: 12px; cursor: pointer; color: white; display: block; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;">Powered By</div><div id="contextMenusPrimisLogo" style="position: relative; float: right; height: 25px; width: 56px; margin-right: 4px; display: block; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" viewBox="0 0 877.5 228.7" style="width: 100%; height: 100%; cursor: pointer;" xml:space="preserve"><path style="fill:#FFFFFF;" d="M154.2,195H115l29.8-146.3H184l-2.4,11.5c5.4-7.8,16-12.8,29-12.8c24,0,47.4,18.8,47.4,49.4 c0,34.9-30.8,62.4-61.9,62.4c-16.5,0-29.5-5-32-12.8L154.2,195z M192.2,124.1c13.6,0,24.5-10.8,24.5-22.7c0-11-8.5-18.8-20.6-18.8 c-14.3,0-24.7,11-24.7,22.5C171.4,116,179.9,124.1,192.2,124.1L192.2,124.1z"></path> <path style="fill:#FFFFFF;" d="M280.2,48.7h40.1l-2.8,13.4c8.7-12.1,18.2-16,29.9-16c4.4,0,8.9,0.4,13.2,1.2l-7.2,35.7 c-4.7-0.8-9.5-1.1-14.3-1.1c-17.3,0-27.9,12.6-32.3,33.1l-8.9,42.7h-40.1L280.2,48.7z"></path> <path style="fill:#FFFFFF;" d="M374.9,48.7h40.1l-22.3,109.1h-40.1L374.9,48.7z M377.7,24c0-12.6,10.8-24,24.5-24c12.1,0,20.6,8.5,20.6,19.9 c0,12.1-10.4,23.8-24.5,23.8C386.1,43.7,377.7,35.5,377.7,24L377.7,24z"></path> <path style="fill:#FFFFFF;" d="M430.8,48.7h40.1l-2.8,13.4c8-12.1,17.1-16,27.9-16c14.3,0,24.9,6.3,29.2,16.9c9.3-11,23.1-17.2,37.4-16.9 c21.4,0,33.6,14.1,33.6,36.2c0,3.6-0.4,7.3-1.1,10.8l-13.4,64.8h-40.1l12.3-61.1c0.2-1.4,0.4-2.9,0.4-4.3c0-7.6-3.7-12.1-10.6-12.1 c-8.7,0-14.3,7.4-16.5,18l-12.3,59.6H475l12.3-61.1c0.2-1.4,0.4-2.9,0.4-4.3c0-7.6-3.7-12.1-10.6-12.1c-8.7,0-14.3,6.9-16.5,17.5 l-12.1,60h-40.1L430.8,48.7z"></path> <path style="fill:#FFFFFF;" d="M619.7,48.7h40.1l-22.3,109.2h-40.1L619.7,48.7z M622.5,24c0-12.6,10.8-24,24.5-24c12.1,0,20.6,8.5,20.6,19.9 c0,12.1-10.4,23.8-24.5,23.8C631,43.7,622.5,35.5,622.5,24z"></path> <path style="fill:#FFFFFF;" d="M676.1,116.7c7.4,8.7,18.4,10.8,29.9,10.8c9.5,0,12.8-2.4,12.8-5.2c0-4.8-18.4-3.9-32.9-11 c-11.5-5.6-17.5-14.5-17.5-26.9c0-21,20.6-39,48.9-39c16.6-0.1,32.7,5.8,45.3,16.7l-20.1,25.3c-7.4-6.7-17.1-10.4-27.1-10.4 c-7.6,0-10.4,2.8-10.4,5.6c0,5.2,15.2,4.8,33.3,11.9c14.3,5.6,18,16.2,18,27.1c0,24.5-18.8,39.4-54.6,39.4 c-22.7,0-40.1-7.4-50.7-19.7L676.1,116.7z"></path> <path style="fill:#FF3E5F;" d="M81.9,195l-6.9,33.8H0L46.5,0h74.9l-6.9,33.8H77.3L44.6,195H81.9z"></path> <path style="fill:#FF3E5F;" d="M795.6,33.7L802.5,0h74.9L831,228.7H756L763,195h37.3l32.6-161.2L795.6,33.7z"></path> </svg></div></div><div id="bottomElements" style="width: 100%; bottom: 0px; left: 0px; position: absolute; opacity: 0; visibility: visible; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; display: none;"><div id="progressBar" style="width: 100%; height: 17px; background-color: rgba(0, 0, 0, 0); bottom: 27px; left: 0px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: none;"><div id="baseBar" style="width: 100%; height: 4px; background-color: rgba(0, 0, 0, 0.6); bottom: 6px; left: 0px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="loadBar" style="width: 100%; height: 4px; background-color: rgba(255, 2, 2, 0.4); bottom: 6px; left: 0px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="timeBar" style="width: 51.6874%; height: 4px; background-color: rgb(255, 2, 2); bottom: 6px; left: 0px; position: absolute; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="timeDot" style="width: 9px; height: 9px; background-color: rgb(255, 2, 2); bottom: 8px; border-radius: 50%; left: 51.6874%; position: absolute; transform: translate(-50%, 50%); z-index: 10; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div style="position:absolute; width:30px; height:30px; top:50%; left:50%; transform:translate(-50%, -50%); cursor: pointer;"></div></div></div><div id="fullScreenBtn" style="width: 20px; height: 20px; bottom: 8px; right: 7px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="fullScreenIco" style="width: 12px; height: 12px; top: 50%; left: 50%; transform: translate(-50%, -50%); position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 18 18" style="pointer-events: none;" xml:space="preserve"> <style type="text/css">   </style>     <g>  <g fill="#fff">   <path d="M17,17h-6v1h6v0h1v-7h-1V17z M1,11H0v7h1v0h6v-1H1V11z M17,0L17,0l-6,0v1h6v6h1V0H17z M1,0H0v7h1V1h6V0L1,0    L1,0z"></path>  </g> </g> </svg></div></div><div id="normalScreenBtn" style="display: none; width: 20px; height: 20px; bottom: 8px; right: 7px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="normalScreenIco" style="width: 12px; height: 12px; top: 50%; left: 50%; transform: translate(-50%, -50%); position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 18 18" style="pointer-events: none;" xml:space="preserve"><style type="text/css"></style><g><g fill="#fff"> <path d="M11,11v1v6h1v-6h6v-1h-6H11z M0,11v1h6v6h1v-6v-1H6H0z M12,6V0h-1v6v1h1h6V6H12z M6,6H0v1h6h1V6V0H6V6z"></path></g></g></svg></div></div><div id="captionsOnBtn" style="width: 20px; height: 20px; bottom: 5px; right: 34px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; display: none;"><div id="captionsOnIco" style="width: 20px; height: 20px; top: 50%; left: 50%; transform: translate(-50%, -50%); position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 29.1 21" style="pointer-events: none;" xml:space="preserve"><g><g fill="#fff"><path d="M9,7c0.4-0.5,0.9-0.7,1.5-0.7c0.7,0,1.2,0.2,1.5,0.5c0.3,0.4,0.5,0.9,0.5,1.7h1.4c0-1.1-0.3-1.9-0.9-2.5 s-1.4-0.8-2.5-0.8S8.6,5.6,7.9,6.3C7.3,7,6.9,8,6.9,9.2v2.9c0,1.3,0.3,2.2,1,2.9c0.7,0.7,1.5,1,2.6,1s1.9-0.3,2.5-0.9 s0.9-1.4,0.9-2.4h-1.4c0,0.8-0.2,1.3-0.5,1.7c-0.3,0.4-0.8,0.5-1.4,0.5c-0.7,0-1.2-0.2-1.5-0.7c-0.4-0.5-0.5-1.2-0.5-2.1v-3 C8.5,8.1,8.6,7.4,9,7z"></path><path d="M17.2,7c0.4-0.5,0.9-0.8,1.5-0.8c0.7,0,1.2,0.2,1.5,0.5c0.3,0.4,0.5,0.9,0.5,1.7h1.4c0-1.1-0.3-1.9-0.9-2.5 s-1.4-0.8-2.5-0.8s-1.9,0.4-2.6,1.1c-0.6,0.7-1,1.7-1,2.9V12c0,1.3,0.3,2.2,1,2.9c0.6,0.7,1.5,1,2.6,1s1.9-0.3,2.5-0.9 s0.9-1.4,0.9-2.4h-1.4c0,0.8-0.2,1.3-0.5,1.7c-0.3,0.4-0.8,0.5-1.4,0.5c-0.7,0-1.2-0.2-1.5-0.7c-0.4-0.5-0.5-1.2-0.5-2.1V9 C16.7,8.1,16.9,7.4,17.2,7z"></path><path d="M0.6,2.5C0.2,2.5,0,2.7,0,3v15c0,0.3,0.2,0.5,0.5,0.5S1,18.3,1,18V3C1.1,2.7,0.9,2.5,0.6,2.5z"></path><path d="M27.1,0h-25C1.7,0,1.5,0.2,1.5,0.5S1.7,1,2,1h25c0.4,0.1,0.6-0.2,0.6-0.5S27.4,0,27.1,0z"></path><path d="M27.1,20h-25c-0.3,0-0.5,0.2-0.5,0.5S1.8,21,2.1,21h25c0.3,0,0.5-0.2,0.5-0.5S27.4,20,27.1,20z"></path><path d="M28.6,2.5c-0.3,0-0.5,0.2-0.5,0.5v15c0,0.3,0.2,0.5,0.5,0.5s0.5-0.2,0.5-0.5V3C29.1,2.7,28.9,2.5,28.6,2.5z"></path></g></g></svg></div></div><div id="captionsOffBtn" style="display: none; width: 20px; height: 20px; bottom: 5px; right: 34px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="captionsOffIco" style="width: 20px; height: 20px; top: 50%; left: 50%; transform: translate(-50%, -50%); position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 29.1 21" style="pointer-events: none;" xml:space="preserve"><g><g fill="#fff"><path d="M28.6,2.5c-0.3,0-0.5,0.2-0.5,0.5v15c0,0.3,0.2,0.5,0.5,0.5c0.3,0,0.5-0.2,0.5-0.5V3 C29.1,2.7,28.9,2.5,28.6,2.5z"></path><path d="M0.5,2.5C0.2,2.5,0,2.7,0,3v15c0,0.3,0.2,0.5,0.5,0.5S1,18.3,1,18V3C1,2.7,0.8,2.5,0.5,2.5z"></path><path d="M13.9,13.2l-2.2,2.6c0.5-0.1,1-0.4,1.3-0.7C13.5,14.6,13.8,14,13.9,13.2z"></path><path d="M10.5,6.3c0.7,0,1.2,0.2,1.5,0.5c0.3,0.4,0.5,0.9,0.5,1.7h1.4c0-1.1-0.3-1.9-0.9-2.5s-1.4-0.8-2.5-0.8 S8.6,5.6,7.9,6.3C7.3,7,6.9,8,6.9,9.2v2.9c0,1.3,0.3,2.2,1,2.9c0.4,0.4,0.8,0.6,1.2,0.8l0.8-1c-0.4-0.1-0.7-0.3-0.9-0.6 c-0.4-0.5-0.5-1.2-0.5-2.1v-3C8.5,8.1,8.6,7.4,9,7C9.4,6.5,9.9,6.3,10.5,6.3z"></path><path d="M21.2,5.9c-0.2-0.2-0.5-0.4-0.8-0.5l-0.8,1c0.2,0.1,0.4,0.2,0.6,0.3c0.3,0.4,0.5,0.9,0.5,1.7h1.4 C22.1,7.3,21.8,6.5,21.2,5.9z"></path><path d="M18.1,5.2c-0.8,0.1-1.4,0.5-2,1c-0.5,0.6-0.9,1.5-1,2.5L18.1,5.2z"></path><path d="M15.1,11.8V12c0,1.3,0.3,2.2,1,2.9c0.6,0.7,1.5,1,2.6,1c1.1,0,1.9-0.3,2.5-0.9c0.6-0.6,0.9-1.4,0.9-2.4h-1.4 c0,0.8-0.2,1.3-0.5,1.7c-0.3,0.4-0.8,0.5-1.4,0.5c-0.7,0-1.2-0.2-1.5-0.7c-0.4-0.5-0.5-1.2-0.5-2.1V9.7L15.1,11.8z"></path><path d="M5.6,20H2.1c-0.3,0-0.5,0.2-0.5,0.5S1.8,21,2.1,21h3.6c-0.1-0.1-0.1-0.2-0.2-0.3C5.5,20.5,5.5,20.3,5.6,20z"></path><path d="M27.1,20H8.2l-0.8,1c0,0,0,0,0,0h19.8c0.3,0,0.5-0.2,0.5-0.5S27.4,20,27.1,20z"></path><path d="M27.1,0h-2.9c0,0,0.1,0.1,0.1,0.1c0.1,0.3,0.1,0.5-0.1,0.8L24.1,1h3c0.3,0,0.5-0.2,0.5-0.5S27.4,0,27.1,0z"></path><path d="M2.1,1h19.5l0.8-0.9c0,0,0,0,0.1-0.1H2.1C1.8,0,1.6,0.2,1.6,0.5S1.8,1,2.1,1z"></path><path d="M23.9,0.3c-0.1-0.1-0.2-0.2-0.3-0.3C23.4,0,23.3,0,23.1,0c-0.2,0-0.3,0.1-0.4,0.3L6,20.1 c-0.1,0.2-0.2,0.3-0.1,0.5c0,0.1,0.1,0.2,0.3,0.3C6.3,21,6.5,21,6.6,21c0.2,0,0.3-0.1,0.4-0.3l16.8-20C24,0.6,24,0.4,23.9,0.3z"></path></g></g></svg></div></div><div id="soundControll" style="bottom: 17px; right: 34px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="soundScrabber" style="position: absolute; right: 0px; bottom: 0px; transform: translate(0px, 50%); display: block; opacity: 0; visibility: hidden; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="soundScrabberBG" style="width: 108px; height: 27px; opacity: 0.72; border-radius: 5px; background-color: rgb(12, 12, 12); position: absolute; right: -8px; bottom: 0px; transform: translate(0px, 50%); user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="soundScrabberHit" style="width: 65px; height: 17px; bottom: 0px; right: 24px; position: absolute; transform: translate(0px, 50%); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="soundScrabberBase" style="width: 100%; height: 4px; bottom: 6px; right: 0px; position: absolute; opacity: 0.34; border-radius: 2px; background-color: rgb(254, 254, 254); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="soundScrabberMain" style="width: 20%; height: 4px; bottom: 6px; left: 0px; position: absolute; border-radius: 2px; background-color: rgb(254, 254, 254); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div></div></div><div id="soundOnBtn" style="width: 16px; height: 16px; bottom: 0px; right: 0px; position: absolute; transform: translate(0px, 50%); display: none; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 489.6 489.6"><g fill="#fff"><path d="M361.1 337.6c2.2 1.5 4.6 2.3 7.1 2.3 3.8 0 7.6-1.8 10-5.2 18.7-26.3 28.5-57.4 28.5-89.9s-9.9-63.6-28.5-89.9c-3.9-5.5-11.6-6.8-17.1-2.9-5.5 3.9-6.8 11.6-2.9 17.1 15.7 22.1 24 48.3 24 75.8 0 27.4-8.3 53.6-24 75.8-3.9 5.4-2.6 13 2.9 16.9zM425.4 396.3c2.2 1.5 4.6 2.3 7.1 2.3 3.8 0 7.6-1.8 10-5.2 30.8-43.4 47.1-94.8 47.1-148.6s-16.3-105.1-47.1-148.6c-3.9-5.5-11.6-6.8-17.1-2.9-5.5 3.9-6.8 11.6-2.9 17.1 27.9 39.3 42.6 85.7 42.6 134.4 0 48.6-14.7 95.1-42.6 134.4-3.9 5.5-2.6 13.1 2.9 17.1zM254.7 415.7c4.3 2.5 9.2 3.8 14.2 3.8 7.4 0 14.4-2.8 19.7-7.9 5.6-5.4 8.7-12.6 8.7-20.4V98.5c0-15.7-12.7-28.4-28.4-28.4-4.9 0-9.8 1.3-14.2 3.8-.3.2-.6.3-.8.5l-100.1 69.2H73.3C32.9 143.6 0 176.5 0 216.9v55.6c0 40.4 32.9 73.3 73.3 73.3h84.5l95.9 69.2c.3.3.7.5 1 .7zm-92.9-94.4H73.3c-26.9 0-48.8-21.9-48.8-48.8v-55.6c0-26.9 21.9-48.8 48.8-48.8h84.3c2.5 0 4.9-.8 7-2.2l102.7-71c.5-.3 1.1-.4 1.6-.4 1.6 0 3.9 1.2 3.9 3.9v292.7c0 1.1-.4 2-1.1 2.8-.7.7-1.8 1.1-2.7 1.1-.5 0-1-.1-1.5-.3l-98.4-71.1c-2.2-1.5-4.7-2.3-7.3-2.3z"></path></g></svg></div><div id="soundOffBtn" style="width: 16px; height: 16px; bottom: 0px; right: 0px; position: absolute; transform: translate(0px, 50%); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; display: block;"><svg xmlns="http://www.w3.org/2000/svg" version="1.1" xmlns:xlink="http://www.w3.org/1999/xlink" preserveAspectRatio="none" x="0px" y="0px" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 612 612"> <defs> <g id="Layer0_0_FILL"> <path fill="#fff" stroke="none" d=" M 371.65 292.8 L 341.1 330.45 341.1 488.8 Q 341.1 490.75 339.7 492.3 338.35 493.75 336.3 493.75 335.15 493.75 334.4 493.3 L 255.4 436.2 252.2 440.15 235.9 460.2 317.1 518.75 Q 317.8 519.35 318.35 519.6 326.7 524.4 336.15 524.4 343.15 524.4 349.45 521.85 355.8 519.3 360.75 514.5 366 509.4 368.85 502.85 371.65 496.25 371.65 489 L 371.65 292.8 M 472.75 193.65 Q 469.1 188.5 462.8 187.45 459.8009765625 186.922265625 457.05 187.5 454.0548828125 188.125390625 451.35 190.05 446.2 193.75 445.15 200 444.9408203125 201.255078125 444.9 202.45 444.8271484375 207.3259765625 447.75 211.45 462.45 232.15 470.1 256.3 477.75 280.45 477.75 306.25 477.75 312.15 477.35 318 476 337.45 470.1 356.05 462.45 380.15 447.75 401 444.1 406 445.15 412.2 446.2 418.35 451.35 422 455.6 424.95 460.25 424.95 463.9 424.95 467.2 423.3 470.55 421.55 472.75 418.45 490.25 393.8 499.35 365.2 506.6 342.25 508 318 508.35 312.05 508.35 306.05 508.35 275.55 499.25 246.95 490.2 218.35 472.75 193.65 M 361.25 98.05 Q 350.85 87.65 336.15 87.65 326.75 87.65 318.35 92.45 317.6 92.85 317.35 93.05 L 192.25 179.5 91.55 179.5 Q 72.95 179.5 55.95 186.7 39.5 193.7 26.85 206.4 14.15 219.1 7.2 235.55 0 252.55 0 271.15 L 0 340.65 Q 0 359.25 7.2 376.25 14.15 392.7 26.85 405.4 39.5 418.1 55.95 425.05 72.95 432.25 91.55 432.25 L 139.65 432.25 164.45 401.65 91.55 401.65 Q 66.4 401.65 48.45 383.75 30.55 365.85 30.55 340.65 L 30.55 271.15 Q 30.55 245.95 48.45 228.05 66.35 210.15 91.55 210.15 L 197 210.15 Q 201.6 210.15 205.8 207.35 L 334.2 118.55 Q 334.95 118.05 336.2 118.05 337.95 118.05 339.4 119.25 341.1 120.6 341.1 122.95 L 341.1 183.7 371.65 146.05 371.65 123.15 Q 371.65 108.4 361.25 98.05 M 558 127.4 Q 555.6 123.8 553.05 120.2 549.35 115.05 543.1 114 536.8 112.95 531.65 116.6 526.5 120.25 525.45 126.55 524.35 132.8 528.05 138 546 163.2 558 191.2 563.55 204.1 567.8 217.6 581.35 260.3 581.35 306 581.35 351.6 567.8 394.4 554.2 437.25 528.05 474 524.45 479.2 525.55 485.45 526.6 491.7 531.75 495.45 536 498.35 540.7 498.35 544.35 498.35 547.65 496.7 551 494.95 553.2 491.85 582 451.05 597 403.75 612 356.35 612 306 612 255.65 596.95 208.3 583.25 165.25 558 127.4 M 513.6 10.4 Q 507.15 8.05 500.5 9.7 493.1 11.55 488.8 17.85 L 37.15 574.05 Q 32.15 581.4 33.85 588.05 35.3 593.95 41.15 597.25 46.9 600.5 53.25 599.4 60.2 598.25 64.2 592.3 L 520.9 29.05 Q 524.85 23 522.3 17.65 519.95 12.75 513.6 10.4 Z"></path> </g> </defs> <g transform="matrix( 1, 0, 0, 1, 0,0) "> <use xlink:href="#Layer0_0_FILL"></use> </g> </svg></div></div><div id="progressTxt" style="pointer-events: auto; bottom: 10px; left: 11px; position: absolute; font-size: 11px; font-family: Roboto; color: rgb(255, 255, 255); user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0);">00:04/00:08</div><div id="skipXsec" style="width: 75px; height: 14px; position: absolute; bottom: 8px; left: 50%; transform: translateX(-50%); user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; display: block;"><div id="skipXsecBackBtn" style="width: 14px; height: 14px; left: 0px; position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="skipXsecBackBtnIco" style="width: 14px; height: 14px; position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" enable-background="new 0 0 512.004 512.004" viewBox="0 0 512.004 512.004" style="pointer-events: none;"><path fill="#fff" d="M464.004 98.346h-368v32h368c14.544 0 15.92 15.92 16 20.688v128c0 3.072-.448 7.152-2.608 9.056-2.816 2.496-8.784 2.624-11.072 2.416-.768-.096-1.552-.16-2.32-.16H54.628l68.688-68.688-22.624-22.624-96 96c-6.256 6.256-6.256 16.368 0 22.624l96 96 22.624-22.624-68.688-68.688h408.464c1.168.112 2.896.24 5.024.24 7.536 0 20.064-1.52 30.144-10.224 6.272-5.424 13.744-15.6 13.744-33.328v-128c0-21.168-12.784-52.688-48-52.688z"></path></svg></div></div><div id="skipXsecForwardBtn" style="width: 14px; height: 14px; right: 0px; position: absolute; cursor: pointer; transform: scaleX(-1); user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="skipXsecForwardBtnIco" style="width: 14px; height: 14px; position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" enable-background="new 0 0 512.004 512.004" viewBox="0 0 512.004 512.004" style="pointer-events: none;"><path fill="#fff" d="M464.004 98.346h-368v32h368c14.544 0 15.92 15.92 16 20.688v128c0 3.072-.448 7.152-2.608 9.056-2.816 2.496-8.784 2.624-11.072 2.416-.768-.096-1.552-.16-2.32-.16H54.628l68.688-68.688-22.624-22.624-96 96c-6.256 6.256-6.256 16.368 0 22.624l96 96 22.624-22.624-68.688-68.688h408.464c1.168.112 2.896.24 5.024.24 7.536 0 20.064-1.52 30.144-10.224 6.272-5.424 13.744-15.6 13.744-33.328v-128c0-21.168-12.784-52.688-48-52.688z"></path></svg></div></div><div id="skipXsecStaticNum" style="position: absolute; left: 50%; transform: translateX(-50%); color: rgb(255, 255, 255); font-size: 10px; line-height: 12px; font-family: Roboto; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;">10 Sec</div></div></div><div id="centerElements" style="top: 50%; left: 50%; transform: translate(-50%, -50%); position: absolute; opacity: 0; visibility: visible; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; display: none;"><div id="playBtn" style="visibility: visible; width: 40px; height: 40px; top: 0px; left: 0px; position: absolute; transform: translate(-50%, -50%); z-index: 1; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; display: none;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 512 512"><path fill="#fff" d="M256 0C114.617 0 0 114.615 0 256s114.617 256 256 256 256-114.615 256-256S397.383 0 256 0zm88.48 269.57l-128 80a16.008 16.008 0 0 1-16.238.422A15.994 15.994 0 0 1 192 336V176c0-5.82 3.156-11.172 8.242-13.992a15.957 15.957 0 0 1 16.238.422l128 80c4.676 2.93 7.52 8.055 7.52 13.57s-2.844 10.641-7.52 13.57z"></path></svg></div><div id="pauseBtn" style="visibility: visible; width: 40px; height: 40px; top: 0px; left: 0px; position: absolute; transform: translate(-50%, -50%); z-index: 1; display: block; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 300.003 300.003"><path fill="#fff" d="M150.001 0c-82.838 0-150 67.159-150 150 0 82.838 67.162 150.003 150 150.003 82.843 0 150-67.165 150-150.003 0-82.841-67.155-150-150-150zM134.41 194.538c0 9.498-7.7 17.198-17.198 17.198s-17.198-7.7-17.198-17.198V105.46c0-9.498 7.7-17.198 17.198-17.198s17.198 7.7 17.198 17.198v89.078zm64.545 0c0 9.498-7.701 17.198-17.198 17.198-9.498 0-17.198-7.7-17.198-17.198V105.46c0-9.498 7.7-17.198 17.198-17.198s17.198 7.7 17.198 17.198v89.078z" class="active-path" data-old_color="#ffffff" data-original="#000000"></path></svg></div><div id="backBtn" style="width: 44px; height: 44px; top: 50%; right: 6px; transform: translate(-50%, -50%); position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="backBtnIco" style="width: 18px; height: 18px; top: 50%; left: 50%; transform: translate(-50%, -50%); position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 42 42"><path fill="#fff" d="M35.965.114a1.005 1.005 0 0 0-1.033.063L7.5 19.095V1a1 1 0 1 0-2 0v40a1 1 0 1 0 2 0V22.905l27.432 18.919a1.012 1.012 0 0 0 1.033.062A1 1 0 0 0 36.5 41V1a1 1 0 0 0-.535-.886z"></path></svg></div></div><div id="nextBtn" style="width: 44px; height: 44px; top: 50%; left: 6px; transform: translate(50%, -50%); position: absolute; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="nextBtnIco" style="width: 18px; height: 18px; top: 50%; left: 50%; transform: scale(-1) translate(50%, 50%); position: relative; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 42 42"><path fill="#fff" d="M35.965.114a1.005 1.005 0 0 0-1.033.063L7.5 19.095V1a1 1 0 1 0-2 0v40a1 1 0 1 0 2 0V22.905l27.432 18.919a1.012 1.012 0 0 0 1.033.062A1 1 0 0 0 36.5 41V1a1 1 0 0 0-.535-.886z"></path></svg></div></div></div><div id="viewsElements" style="top: 32px; left: 9px; position: absolute; z-index: 0; display: block; opacity: 0; visibility: visible; font-family: Roboto; pointer-events: auto; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0);"><div id="viewsIcon" style="position: relative; width: 21px; height: 16px; cursor: pointer; pointer-events: auto; color: white; float: left; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); bottom: 0px;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width:100%; height:100%;" viewBox="0 0 511.999 511.999"> <path fill="#fff" d="M508.745 246.041c-4.574-6.257-113.557-153.206-252.748-153.206S7.818 239.784 3.249 246.035a16.896 16.896 0 0 0 0 19.923c4.569 6.257 113.557 153.206 252.748 153.206s248.174-146.95 252.748-153.201a16.875 16.875 0 0 0 0-19.922zM255.997 385.406c-102.529 0-191.33-97.533-217.617-129.418 26.253-31.913 114.868-129.395 217.617-129.395 102.524 0 191.319 97.516 217.617 129.418-26.253 31.912-114.868 129.395-217.617 129.395z"></path> <path fill="#fff" d="M255.997 154.725c-55.842 0-101.275 45.433-101.275 101.275s45.433 101.275 101.275 101.275S357.272 311.842 357.272 256s-45.433-101.275-101.275-101.275zm0 168.791c-37.23 0-67.516-30.287-67.516-67.516s30.287-67.516 67.516-67.516 67.516 30.287 67.516 67.516-30.286 67.516-67.516 67.516z"></path> </svg></div><div id="viewsNum" style="position: relative; width: max-content; height: 20px; cursor: pointer; pointer-events: auto; color: white; float: left; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); font-size: 13px; left: 0px;">127.3K</div><div id="likesIcon" style="position: relative; width: 16px; height: 16px; cursor: pointer; float: left; left: 10px; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; bottom: 2px;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" viewBox="0 0 850 768" style="pointer-events: none; width:100%; height:100%;"> <path fill="#fff" d="M775.9,522.1c15.7-20.1,23.3-41.6,22.3-63.8c-1-24.4-11.9-43.5-20.9-55.2c10.4-26,14.5-67-20.4-98.8c-25.5-23.3-68.9-33.7-129-30.8c-42.2,1.9-77.6,9.8-79,10.1h-0.2c-8,1.4-16.5,3.2-25.2,5.1c-0.6-10.3,1.1-35.8,20.1-93.3c22.5-68.4,21.2-120.8-4.2-155.8C512.8,2.9,470.2,0,457.7,0c-12,0-23.1,5-31,14.1c-17.8,20.7-15.7,58.9-13.5,76.6c-21.2,56.9-80.6,196.3-130.9,235c-1,0.6-1.8,1.4-2.6,2.2c-14.8,15.6-24.7,32.4-31.5,47.2c-9.5-5.1-20.2-8-31.8-8h-98c-36.9,0-66.8,30-66.8,66.8v261c0,36.9,30,66.8,66.8,66.8h98c14.3,0,27.6-4.5,38.5-12.2l37.7,4.5c5.8,0.8,108.6,13.8,214.1,11.7c19.1,1.4,37.1,2.2,53.8,2.2c28.7,0,53.8-2.2,74.7-6.7c49.1-10.4,82.7-31.3,99.7-62c13-23.4,13-46.7,10.9-61.5c32-28.9,37.6-60.9,36.5-83.4C781.8,541.4,778.9,530.3,775.9,522.1L775.9,522.1z M118.5,718.4c-13,0-23.4-10.6-23.4-23.4V433.8c0-13,10.6-23.4,23.4-23.4h98c13,0,23.4,10.6,23.4,23.4v261c0,13-10.6,23.4-23.4,23.4h-98V718.4L118.5,718.4z M734.8,503.3c-6.7,7.1-8,17.8-2.9,26.2c0,0.2,6.6,11.4,7.4,26.8c1.1,21-9,39.7-30.2,55.6c-7.5,5.8-10.6,15.7-7.4,24.7c0,0.2,6.9,21.4-4.3,41.4c-10.8,19.3-34.7,33.1-71,40.8c-29.1,6.3-68.6,7.4-117.1,3.5H507c-103.3,2.2-207.7-11.2-208.8-11.4h-0.2l-16.2-1.9c1-4.5,1.4-9.3,1.4-14.1V433.8c0-6.9-1.1-13.7-3.1-19.9c2.9-10.8,10.9-34.7,29.9-55.1c72.1-57.2,142.6-250.1,145.7-258.4c1.3-3.4,1.6-7.1,1-10.8c-2.7-18-1.8-40,2.1-46.6c8.5,0.2,31.5,2.6,45.3,21.7c16.4,22.6,15.7,63.1-1.9,116.8c-27,81.7-29.2,124.8-7.9,143.7c10.6,9.5,24.7,10,35,6.3c9.8-2.2,19.1-4.2,27.9-5.6c0.6-0.2,1.4-0.3,2.1-0.5C608.7,314.6,697,308,727.7,336c26,23.8,7.5,55.2,5.5,58.6c-5.9,9-4.2,20.7,3.9,27.9c0.2,0.2,17,16.1,17.8,37.4C755.5,474.3,748.7,488.9,734.8,503.3L734.8,503.3z"></path> <path fill="#000" fill-opacity="0" d="M118.5,718.4c-13,0-23.4-10.6-23.4-23.4V433.8c0-13,10.6-23.4,23.4-23.4h98c13,0,23.4,10.6,23.4,23.4v261c0,13-10.6,23.4-23.4,23.4h-98V718.4L118.5,718.4z M734.8,503.3c-6.7,7.1-8,17.8-2.9,26.2c0,0.2,6.6,11.4,7.4,26.8c1.1,21-9,39.7-30.2,55.6c-7.5,5.8-10.6,15.7-7.4,24.7c0,0.2,6.9,21.4-4.3,41.4c-10.8,19.3-34.7,33.1-71,40.8c-29.1,6.3-68.6,7.4-117.1,3.5H507c-103.3,2.2-207.7-11.2-208.8-11.4h-0.2l-16.2-1.9c1-4.5,1.4-9.3,1.4-14.1V433.8c0-6.9-1.1-13.7-3.1-19.9c2.9-10.8,10.9-34.7,29.9-55.1c72.1-57.2,142.6-250.1,145.7-258.4c1.3-3.4,1.6-7.1,1-10.8c-2.7-18-1.8-40,2.1-46.6c8.5,0.2,31.5,2.6,45.3,21.7c16.4,22.6,15.7,63.1-1.9,116.8c-27,81.7-29.2,124.8-7.9,143.7c10.6,9.5,24.7,10,35,6.3c9.8-2.2,19.1-4.2,27.9-5.6c0.6-0.2,1.4-0.3,2.1-0.5C608.7,314.6,697,308,727.7,336c26,23.8,7.5,55.2,5.5,58.6c-5.9,9-4.2,20.7,3.9,27.9c0.2,0.2,17,16.1,17.8,37.4C755.5,474.3,748.7,488.9,734.8,503.3L734.8,503.3z"></path> </svg></div><div id="likesNum" style="position: relative; width: max-content; height: 20px; cursor: pointer; pointer-events: auto; color: white; float: left; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); font-size: 13px; left: 14px; display: none;">2</div></div><div id="topLeftElements" style="top: 0px; left: 0px; width: calc(100% - 90px); position: absolute; opacity: 1; visibility: visible; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="titleObj" style="top: 10px; left: 11px; width: 100%; font-family: Roboto; font-size: 14px; color: white; position: absolute; display: inline-flex; opacity: 0; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: none;"><div id="titleTxt" style="cursor: pointer; padding-right: 15px; text-overflow: ellipsis; white-space: nowrap; overflow: hidden; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;">Bible Verse_ Psalm 118_24</div><div id="titleIco" style="pointer-events: auto; margin-left: -15px; width: 12px; min-width: 12px; height: 12px; padding-right: 5px; transform: scale(-1, 1); margin-top: 1px; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); display: none;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width:100%; height:100%;" viewBox="-22 0 512 512.00667"><path fill="#fff" d="M422.188 98.813h-44.91a8.986 8.986 0 0 0-8.985 8.984c0 4.957 4.023 8.98 8.984 8.98h44.91c14.856 0 26.95 12.09 26.95 26.95v323.367c0 14.86-12.094 26.949-26.95 26.949H98.82c-14.86 0-26.949-12.09-26.949-26.95V251.517c0-4.957-4.023-8.985-8.98-8.985-4.961 0-8.985 4.028-8.985 8.985v215.578c0 24.765 20.149 44.914 44.914 44.914h323.368c24.765 0 44.914-20.149 44.914-44.914V143.727c0-24.766-20.149-44.914-44.914-44.914zm0 0"></path><path fill="#fff" d="M156.566 250.645a8.971 8.971 0 0 0 5.121-8.114v-53.894h8.985c108.531 0 156.437 53.07 198.539 138.699a8.974 8.974 0 0 0 10.094 4.789 8.98 8.98 0 0 0 6.945-8.75c0-228.938-103.613-260.492-215.578-260.492h-8.984V8.988a8.979 8.979 0 0 0-5.122-8.11 8.939 8.939 0 0 0-9.52 1.142L3.329 118.79c-2.101 1.698-3.324 4.269-3.324 6.972s1.223 5.261 3.324 6.968l143.719 116.774a8.942 8.942 0 0 0 9.52 1.14zm0 0"></path></svg></div></div></div><div id="topRightElements" style="top: 0px; right: 0px; position: absolute; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; visibility: visible;"><div id="primisLogoWrapper" style="transform: scale(1.1); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="primisLogo" style="top: 8px; right: 7px; width: 12px; position: absolute; opacity: 1; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: none; display: none;"><div id="pl1" style=" height: 15px; width: 6px;  position:absolute; top:0; left:0; pointer-events:none;"><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 121.4 228.7" style="enable-background:new 0 0 121.4 228.7;" xml:space="preserve"><style type="text/css">	.stlp0{fill:#FF3E5F;}</style><path class="stlp0" d="M81.9,195L75,228.8H0L46.5,0h74.9l-6.9,33.8H77.3L44.6,195H81.9z"></path></svg></div><div id="primisLogoTxt" style="overflow: hidden; margin-right: 5px; position: relative; height: 15px; opacity: 0; pointer-events: none;"><div id="pl2" style=" height: 15px; width: 37px;  position:absolute; top:0; left:0;"><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 762.6 228.7" style="enable-background:new 0 0 762.6 228.7;" xml:space="preserve"><style type="text/css">	.st1{fill:#fffffe;}</style><path class="st1" d="M154.2,195H115l29.8-146.3H184l-2.4,11.5c5.4-7.8,16-12.8,29-12.8c24,0,47.4,18.8,47.4,49.4	c0,34.9-30.8,62.4-61.9,62.4c-16.5,0-29.5-5-32-12.8L154.2,195z M192.2,124.1c13.6,0,24.5-10.8,24.5-22.7c0-11-8.5-18.8-20.6-18.8	c-14.3,0-24.7,11-24.7,22.5C171.4,116,179.9,124.1,192.2,124.1L192.2,124.1z"></path><path class="st1" d="M280.2,48.7h40.1l-2.8,13.4c8.7-12.1,18.2-16,29.9-16c4.4,0,8.9,0.4,13.2,1.2L353.4,83	c-4.7-0.8-9.5-1.1-14.3-1.1c-17.3,0-27.9,12.6-32.3,33.1l-8.9,42.7h-40.1L280.2,48.7z"></path><path class="st1" d="M374.9,48.7H415l-22.3,109.1h-40.1L374.9,48.7z M377.7,24c0-12.6,10.8-24,24.5-24c12.1,0,20.6,8.5,20.6,19.9	c0,12.1-10.4,23.8-24.5,23.8C386.1,43.7,377.7,35.5,377.7,24L377.7,24z"></path><path class="st1" d="M430.8,48.7h40.1l-2.8,13.4c8-12.1,17.1-16,27.9-16c14.3,0,24.9,6.3,29.2,16.9c9.3-11,23.1-17.2,37.4-16.9	c21.4,0,33.6,14.1,33.6,36.2c0,3.6-0.4,7.3-1.1,10.8l-13.4,64.8h-40.1l12.3-61.1c0.2-1.4,0.4-2.9,0.4-4.3c0-7.6-3.7-12.1-10.6-12.1	c-8.7,0-14.3,7.4-16.5,18L514.9,158H475l12.3-61.1c0.2-1.4,0.4-2.9,0.4-4.3c0-7.6-3.7-12.1-10.6-12.1c-8.7,0-14.3,6.9-16.5,17.5	l-12.1,60h-40.1L430.8,48.7z"></path><path class="st1" d="M619.7,48.7h40.1l-22.3,109.2h-40.1L619.7,48.7z M622.5,24c0-12.6,10.8-24,24.5-24c12.1,0,20.6,8.5,20.6,19.9	c0,12.1-10.4,23.8-24.5,23.8C631,43.7,622.5,35.5,622.5,24z"></path><path class="st1" d="M676.1,116.7c7.4,8.7,18.4,10.8,29.9,10.8c9.5,0,12.8-2.4,12.8-5.2c0-4.8-18.4-3.9-32.9-11	c-11.5-5.6-17.5-14.5-17.5-26.9c0-21,20.6-39,48.9-39c16.6-0.1,32.7,5.8,45.3,16.7l-20.1,25.3c-7.4-6.7-17.1-10.4-27.1-10.4	c-7.6,0-10.4,2.8-10.4,5.6c0,5.2,15.2,4.8,33.3,11.9c14.3,5.6,18,16.2,18,27.1c0,24.5-18.8,39.4-54.6,39.4	c-22.7,0-40.1-7.4-50.7-19.7L676.1,116.7z"></path></svg></div></div><div id="pl3" style=" height: 15px; width: 6px;  position:absolute; top:0; right:0; "><svg version="1.1" id="Layer_1" xmlns="http://www.w3.org/2000/svg" xmlns:xlink="http://www.w3.org/1999/xlink" x="0px" y="0px" viewBox="0 0 121.5 228.7" style="enable-background:new 0 0 121.5 228.7;" xml:space="preserve"><style type="text/css">	.st2{fill:#FF3E5F;}</style><path class="st2" d="M39.6,33.7L46.5,0h74.9L75,228.7H0L7,195h37.3L76.9,33.8L39.6,33.7z"></path></svg></div></div></div><div id="shareContainer" style="top: 26px; right: 8px; position: absolute; z-index: 1; opacity: 0; display: block; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: none;"><div id="shareBG" style="top: 0px; right: 0px; width: 242px; height: 34px; object-fit: contain; opacity: 0; border-radius: 5px; background-color: rgb(12, 12, 12); position: absolute; display: none; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="shareBtn" style="width: 14px; height: 14px; position: absolute; top: 10px; right: 10px; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 481.6 481.6"><path fill="#fff" d="M381.6 309.4c-27.7 0-52.4 13.2-68.2 33.6l-132.3-73.9c3.1-8.9 4.8-18.5 4.8-28.4 0-10-1.7-19.5-4.9-28.5l132.2-73.8c15.7 20.5 40.5 33.8 68.3 33.8 47.4 0 86.1-38.6 86.1-86.1S429 0 381.5 0s-86.1 38.6-86.1 86.1c0 10 1.7 19.6 4.9 28.5l-132.1 73.8c-15.7-20.6-40.5-33.8-68.3-33.8-47.4 0-86.1 38.6-86.1 86.1s38.7 86.1 86.2 86.1c27.8 0 52.6-13.3 68.4-33.9l132.2 73.9c-3.2 9-5 18.7-5 28.7 0 47.4 38.6 86.1 86.1 86.1s86.1-38.6 86.1-86.1-38.7-86.1-86.2-86.1zm0-282.3c32.6 0 59.1 26.5 59.1 59.1s-26.5 59.1-59.1 59.1-59.1-26.5-59.1-59.1 26.6-59.1 59.1-59.1zM100 299.8c-32.6 0-59.1-26.5-59.1-59.1s26.5-59.1 59.1-59.1 59.1 26.5 59.1 59.1-26.6 59.1-59.1 59.1zm281.6 154.7c-32.6 0-59.1-26.5-59.1-59.1s26.5-59.1 59.1-59.1 59.1 26.5 59.1 59.1-26.5 59.1-59.1 59.1z"></path></svg></div><div id="facebookBtn" style="top: 10px; right: 44px; width: 14px; height: 14px; position: absolute; display: none; opacity: 0; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="-110 1 511 511.99996"><path fill="#fff" d="M180 512H98.008c-13.695 0-24.836-11.14-24.836-24.836V302.227H25.336C11.64 302.227.5 291.082.5 277.39v-79.246c0-13.696 11.14-24.836 24.836-24.836h47.836v-39.684c0-39.348 12.355-72.824 35.726-96.805C132.375 12.73 165.184 0 203.778 0l62.53.102c13.672.023 24.794 11.164 24.794 24.835v73.579c0 13.695-11.137 24.836-24.829 24.836l-42.101.015c-12.84 0-16.11 2.574-16.809 3.363-1.152 1.31-2.523 5.008-2.523 15.223v31.352h58.27c4.386 0 8.636 1.082 12.288 3.12 7.88 4.403 12.778 12.727 12.778 21.723l-.031 79.247c0 13.687-11.141 24.828-24.836 24.828h-58.47v184.941C204.84 500.86 193.696 512 180 512zm-76.813-30.016h71.633V288.79c0-9.144 7.442-16.582 16.582-16.582h66.727l.027-68.883h-66.758c-9.14 0-16.578-7.437-16.578-16.582v-44.789c0-11.726 1.192-25.062 10.043-35.086 10.696-12.117 27.551-13.515 39.301-13.515l36.922-.016V30.109l-57.332-.093c-62.024 0-100.566 39.703-100.566 103.609v53.117c0 9.14-7.438 16.582-16.579 16.582H30.516v68.883h56.093c9.141 0 16.579 7.438 16.579 16.582zM266.25 30.117h.004zm0 0"></path></svg></div><div id="twitterBtn" style="top: 10px; right: 78px; width: 14px; height: 14px; position: absolute; display: none; opacity: 0; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 511.999 511.999">    <path fill="#fff" d="M509.738 109.538a15.263 15.263 0 0 0-14.332-7.215l-25.53 2.224 24.403-49.193a15.268 15.268 0 1 0-20.393-20.498l-64.587 31.631c-39.903-21.066-89.756-14.813-124.06 16.436-28.63 26.08-43.679 66.187-40.873 106.183-74.829-7.5-138.169-50.331-175.623-119.537a15.269 15.269 0 0 0-25.7-1.817c-23.23 31.382-24.86 74.41-6.649 109.809-5.02-1.268-10.362-2.984-16.286-4.963A15.266 15.266 0 0 0 .205 189.564c7.206 43.688 32.682 77.264 72.926 97.138a128.229 128.229 0 0 1-16.024 4.44 15.267 15.267 0 0 0-8.034 25.378c28.981 30.978 70.845 46.225 100.581 53.539-33.81 26.477-70.307 30.908-123.341 29.087-6.139-.186-11.795 3.272-14.376 8.837a15.27 15.27 0 0 0 2.544 16.681c22.108 24.369 102.654 51.847 187.326 53.907 3.349.081 6.765.125 10.253.125 59.451-.001 138.022-12.745 194.419-69.142 42.687-42.686 69.387-91.827 79.356-146.053 8.052-43.797 2.963-78.869.518-95.725-.18-1.239-.372-2.554-.528-3.704l23.881-38.49a15.267 15.267 0 0 0 .032-16.044zm-53.606 62.622c4.531 31.224 18.316 126.24-71.245 215.801-52.06 52.059-127.888 61.407-182.338 60.084-46.461-1.13-88.019-10.016-118.139-20.15 12.229-1.5 23.127-3.814 33.414-7.011 28.359-8.815 52.317-24.676 75.395-49.915a15.268 15.268 0 0 0-9.327-25.447c-14.559-1.866-58.305-9.23-92.753-32.992 9.927-4.06 19.501-9.141 28.643-15.205a15.27 15.27 0 0 0-4.569-27.494c-28.195-7.392-62.223-23.932-77.914-60.855 10.02 1.756 20.915 2.181 32.535-.424a15.268 15.268 0 0 0 7.645-25.504c-19.524-20.222-26.825-49.364-20.561-74.301 20.514 29.981 46.503 55.213 76.279 73.868 38.091 23.865 82.549 37.054 128.57 38.137a15.32 15.32 0 0 0 12.157-5.573 15.27 15.27 0 0 0 3.107-13.007c-7.918-35.595 3.375-73.541 28.771-96.674 26.34-23.995 65.269-27.635 94.666-8.846a15.27 15.27 0 0 0 14.938.846l31.541-15.448-16.509 33.281a15.265 15.265 0 0 0 1.065 15.388 15.252 15.252 0 0 0 13.938 6.607l22.242-1.937-10.246 16.515c-3.439 5.543-2.726 10.461-1.305 20.256z"></path></svg></div><div id="linkedinBtn" style="top: 10px; right: 112px; width: 14px; height: 14px; position: absolute; display: none; opacity: 0; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 510 510">  <path fill="#fff" stroke="none" d=" M 29.4 170.55 Q 25.0033203125 174.9400390625 25 181.15 L 25 478.95 Q 25.0033203125 485.1599609375 29.4 489.55 33.7900390625 493.9466796875 40 493.95 L 132.7 493.95 Q 138.9099609375 493.9466796875 143.3 489.55 147.6966796875 485.1599609375 147.7 478.95 L 147.7 181.15 Q 147.6966796875 174.9400390625 143.3 170.55 138.9099609375 166.1533203125 132.7 166.15 L 40 166.15 Q 33.7900390625 166.1533203125 29.4 170.55 M 55 196.15 L 117.7 196.15 117.7 463.95 55 463.95 55 196.15 M 134.85 38.35 L 134.85 38.35 Q 114.7173828125 18.1341796875 86.3 18.15 57.823828125 18.1466796875 37.7 38.35 L 37.7 38.35 Q 17.70234375 58.46015625 17.7 86.85 17.694921875 115.2359375 37.7 135.35 L 37.7 135.4 Q 57.82109375 155.55390625 86.3 155.55 114.702734375 155.55390625 134.85 135.4 L 134.85 135.4 Q 154.954296875 115.2505859375 154.95 86.85 154.9470703125 58.4404296875 134.85 38.35 M 113.6 59.55 L 113.65 59.55 Q 124.9529296875 70.8595703125 124.95 86.85 124.9453125 102.8494140625 113.6 114.2 L 113.65 114.2 Q 102.297265625 125.54609375 86.3 125.55 70.296875 125.54609375 59 114.2 L 58.95 114.2 Q 47.7052734375 102.846484375 47.7 86.85 47.69765625 70.857421875 58.95 59.55 L 59 59.5 Q 70.2935546875 48.1533203125 86.3 48.15 102.2828125 48.166015625 113.6 59.55 M 366.55 287.35 L 366.55 287.35 Q 369.8208984375 295.9189453125 370.95 308.6 371.6982421875 317.625 371.7 334.2 L 371.7 479.05 Q 371.7033203125 485.2599609375 376.1 489.65 380.4900390625 494.0466796875 386.7 494.05 L 479.35 494.05 Q 485.5599609375 494.0466796875 489.95 489.65 494.3466796875 485.2599609375 494.35 479.05 L 494.35 315.75 Q 494.346484375 281.159765625 489.9 257 484.36796875 226.735546875 471.6 206.7 442.5888671875 158.8494140625 368.25 158.85 332.56171875 158.936328125 304.3 176.4 L 304.3 176.35 Q 299.108203125 179.589453125 294.45 183.15 L 294.45 181.15 Q 294.4466796875 174.9400390625 290.05 170.55 285.6599609375 166.1533203125 279.45 166.15 L 190.75 166.15 Q 184.5400390625 166.1533203125 180.15 170.55 175.7533203125 174.9400390625 175.75 181.15 L 175.75 479.05 Q 175.7533203125 485.2599609375 180.15 489.65 184.5400390625 494.0466796875 190.75 494.05 L 283.15 494.05 Q 289.3599609375 494.0466796875 293.75 489.65 298.1466796875 485.2599609375 298.15 479.05 L 298.15 331.75 Q 298.1484375 301.451171875 305.95 286.25 L 306 286.2 Q 315.576171875 270.294921875 338.65 270.3 359.621484375 269.85 366.55 287.35 M 446.2 222.7 Q 456.0318359375 238.5634765625 460.4 262.4 464.353515625 284.340625 464.35 315.75 L 464.35 464.05 401.7 464.05 401.7 334.2 Q 401.7017578125 315.975 400.85 306.05 L 400.85 306 Q 399.27265625 288.264453125 394.45 276.35 L 394.45 276.3 Q 380.37734375 240.7470703125 338.65 240.3 296.2140625 240.305078125 279.7 271.75 L 279.7 271.75 Q 268.1515625 291.6453125 268.15 331.75 L 268.15 464.05 205.75 464.05 205.75 196.15 264.45 196.15 264.45 221.9 Q 264.4533203125 228.1099609375 268.85 232.5 273.2400390625 236.8966796875 279.45 236.9 L 280.65 236.9 Q 284.8103515625 236.90390625 288.35 234.75 291.9462890625 232.5927734375 293.9 228.9 302.626171875 212.6734375 320.05 201.9 L 320.05 201.9 Q 341.3384765625 188.7634765625 368.25 188.85 423.7208984375 188.8505859375 446.2 222.7 Z"></path> </svg> </div><div id="pinterestBtn" style="top: 10px; right: 146px; width: 14px; height: 14px; position: absolute; display: none; opacity: 0; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 520 520"> <path fill="#fff" stroke="none" d=" M 340.95 115.95 L 341 115.95 Q 366.8947265625 141.594921875 366.9 183.05 366.915234375 211.75390625 360.8 237.75 354.941796875 262.813671875 344.05 281.85 L 344 281.9 Q 333.6970703125 299.9265625 319.95 309.8 L 319.9 309.8 Q 306.843359375 319.0630859375 291.8 319.15 276.2375 319.3591796875 267 308.4 L 266.95 308.35 Q 258.0572265625 297.5625 261.55 282.85 L 261.55 282.8 Q 264.0701171875 272.1232421875 268.85 256.55 L 268.8 256.55 Q 274.4572265625 238.3400390625 277.05 229.6 L 277.05 229.55 Q 287.273828125 195.3291015625 287.05 178.6 287.04765625 154.2626953125 274.2 139.05 L 274.2 139 Q 259.7958984375 120.948046875 233.7 120.95 203.5919921875 120.5439453125 182.95 147.95 L 182.9 147.95 Q 163.8544921875 172.7755859375 163.85 209.55 163.86328125 228.55703125 169 246.3 L 169 246.35 Q 170.6 251.8517578125 172.2 255.95 L 136.75 406 136.75 406 Q 129.925 434.76171875 131.35 475.2 131.394921875 476.27890625 131.6 477.4 136.5583984375 504.404296875 152.8 508.05 166.89921875 515.028515625 189.1 498.8 191.389453125 497.1279296875 192.9 494.7 215.4203125 459.2328125 223.35 429.8 L 223.35 429.8 Q 225.5431640625 421.843359375 234.7 385.7 L 234.65 385.75 236.6 378.25 Q 242.115234375 382.219140625 248.75 385.75 L 248.8 385.75 Q 272.2091796875 398.1158203125 298.3 398.1 335.6607421875 398.0712890625 367.6 381.65 367.625 381.6623046875 367.65 381.65 398.4412109375 366.0109375 421.1 336.7 443.12890625 308.362890625 454.85 270.2 466.613671875 232.165625 466.65 188.35 466.666796875 153.6826171875 452.5 121.3 L 452.5 121.35 Q 438.5078125 89.2806640625 412.7 64.5 386.191796875 39.1021484375 351.05 25.2 314.0994140625 10.583984375 271.65 10.55 220.0044921875 10.5455078125 177.2 28.05 137.6974609375 44.1279296875 108.65 73.55 L 108.7 73.55 Q 81.7677734375 100.7685546875 67.35 135.6 L 67.35 135.55 Q 53.3587890625 169.0849609375 53.4 203.15 53.3939453125 245.4220703125 69.25 276.45 L 69.25 276.45 Q 87.139453125 311.978515625 120.9 325.6 L 120.8 325.6 Q 132.50859375 330.494140625 141.95 325.4 L 142 325.35 Q 152.0927734375 321.062109375 155.3 308.35 L 155.35 308.3 Q 159.7423828125 290.978125 161.1 285.65 161.11328125 285.6248046875 161.1 285.55 163.7869140625 275.45 161.7 269.65 L 161.7 269.7 Q 160.8388671875 264.51171875 153.95 256.35 146.6896484375 247.687109375 142.95 237.3 L 142.95 237.25 Q 138.7033203125 225.1880859375 138.7 210.25 138.6919921875 186.1400390625 147.4 164.15 156.05078125 142.311328125 172.1 125.65 L 172.05 125.65 Q 188.7466796875 108.4470703125 211.15 99.1 L 211.15 99.1 Q 235.3005859375 89.054296875 263.65 89.1 312.636328125 89.1048828125 340.95 115.95 M 361.9 94.45 L 361.9 94.4 Q 326.141796875 59.094921875 263.65 59.1 229.098828125 59.1458984375 199.65 71.4 L 199.6 71.4 Q 171.452734375 83.152734375 150.5 104.8 L 150.5 104.8 Q 130.3490234375 125.688671875 119.5 153.1 108.7078125 180.359765625 108.7 210.25 108.6966796875 230.8119140625 114.7 247.4 L 114.7 247.4 Q 120.2103515625 262.812890625 131 275.65 132.08984375 276.9224609375 132.7 277.35 L 132.7 277.4 Q 132.23984375 277.4806640625 132.1 278 L 132.05 278.1 Q 130.9310546875 282.5537109375 127.55 295.7 107.378125 285.471875 96.05 262.95 L 96.05 262.9 Q 83.4060546875 237.577734375 83.4 203.15 83.44140625 174.9146484375 95.05 147.1 L 95.05 147.05 Q 107.23203125 117.6318359375 130 94.65 L 130 94.65 Q 154.801953125 69.5720703125 188.5 55.85 226.1953125 40.554296875 271.65 40.55 308.2 40.515625 340 53.1 369.607421875 64.7974609375 391.95 86.15 413.3921875 106.719140625 425 133.35 L 425 133.35 Q 436.6330078125 159.9173828125 436.65 188.35 436.6861328125 227.4341796875 426.2 261.35 416.1708984375 294.0369140625 397.4 318.35 378.9775390625 342.1642578125 354 354.9 L 353.95 354.95 Q 328.31015625 368.1287109375 298.3 368.1 279.590625 368.0841796875 262.8 359.2 L 262.8 359.25 Q 248.23359375 351.52578125 242.15 340.75 240.01015625 336.640625 235.9 334.45 231.778125 332.262890625 227.15 332.8 222.568359375 333.3482421875 219.05 336.4 215.540625 339.491015625 214.35 343.95 L 205.6 378.3 205.6 378.35 Q 196.5568359375 414.055859375 194.4 421.9 L 194.4 421.85 Q 187.5673828125 446.5330078125 169.15 476.15 164.917578125 479.0849609375 162.8 479.75 162.223828125 477.9609375 161.3 473.15 160.1587890625 437.973828125 165.95 412.85 L 202.55 258 Q 203.8064453125 252.7384765625 201.35 247.85 199.6349609375 244.2423828125 197.8 237.95 L 197.8 237.95 Q 193.8365234375 224.242578125 193.85 209.55 193.8453125 183.644921875 206.85 166.05 L 206.85 166 Q 217.9248046875 151.3541015625 233.7 150.95 244.7080078125 150.951953125 251 158.05 L 250.95 158 Q 257.0521484375 166.033984375 257.05 178.6 256.8263671875 192.469921875 248.3 221 L 248.3 221 Q 245.7427734375 229.6599609375 240.15 247.7 L 240.15 247.75 Q 235.0298828125 264.476953125 232.35 275.95 225.2498046875 305.946875 244.1 327.8 L 244.1 327.75 Q 261.9630859375 348.94140625 291.8 349.15 316.1564453125 349.2369140625 337.25 334.3 L 337.25 334.3 Q 356.102734375 321.1734375 370.05 296.8 L 370.05 296.8 Q 383.0583984375 274.2859375 390 244.6 396.884765625 215.3458984375 396.9 183.05 396.9052734375 128.1248046875 361.9 94.45 Z"></path></svg> </div><div id="whatsappBtn" style="top: 10px; right: 180px; width: 14px; height: 14px; position: absolute; display: none; opacity: 0; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 520 520"> <path fill="#fff" stroke="none" d=" M 438.75 79.25 Q 404.5 45 360.1 26.25 314.15 6.8 263.8 6.8 213.45 6.8 167.45 26.25 123.1 45 88.8 79.25 54.55 113.5 35.8 157.9 16.35 203.85 16.35 254.2 16.35 288.75 25.65 321.65 34.7 353.6 51.85 381.95 52.2 382.5 52.3 383.15 L 52.3 383.2 Q 52.35 383.8 52.15 384.45 L 8.8 513.25 141.8 470.7 Q 142.35 470.5 143 470.55 143.6 470.6 144.15 470.9 171.15 485.85 200.95 493.65 231.65 501.65 263.7 501.65 314.05 501.65 360.05 482.2 404.45 463.45 438.75 429.15 473 394.9 491.75 350.5 511.2 304.55 511.2 254.2 511.2 203.85 491.75 157.9 473 113.5 438.75 79.25 M 413.3 104.8 Q 442.55 134.05 458.65 172 475.3 211.3 475.3 254.25 475.25 297.2 458.6 336.45 442.5 374.4 413.2 403.7 383.95 432.95 346 449.05 306.7 465.7 263.8 465.7 232.1 465.7 202.05 456.5 173.6 447.8 148.65 431.5 L 69.9 456.7 69.95 456.7 Q 68.85 457.05 67.8 456.6 L 67.75 456.6 Q 66.7 456.1 66.25 455.05 65.8 454 66.15 452.9 L 91.65 376.95 Q 72.95 350.85 62.95 320.5 52.3 288.4 52.3 254.25 52.3 211.3 68.95 172 85.05 134.05 114.35 104.8 143.6 75.5 181.55 59.4 220.85 42.75 263.8 42.75 306.75 42.75 346.05 59.4 384 75.5 413.3 104.8 M 338.3 288.7 Q 333.05 286.65 330.55 286.85 326.85 287.15 323.8 291.45 320.8 295.6 312.95 304.4 306.5 311.65 302.55 315.7 299.65 318.8 296.55 319.1 293.4 319.35 288.7 316.85 278.9 312.4 270.05 307.25 253.3 297.45 238.5 283.2 226.65 271.8 215.95 256.2 209.85 247.25 204.55 237.6 202.2 233.25 203.65 229.85 204.55 227.7 207.85 224.7 209.75 222.95 212.75 219.6 216.2 215.75 217.85 213.9 220 211.55 221.65 208.95 221.7 208.9 224.7 203.55 227.45 198.55 224.55 192.35 223.45 190.05 206.7 144.15 203.55 135.75 199.3 133.7 198 133.1 192.7 132.95 190.75 132.9 186.8 132.5 182.5 132.05 180.5 131.95 176.6 131.75 172.5 133.25 167.3 135.15 163.2 139.35 153.35 149.3 149.15 156.1 139.55 171.65 138.8 191.85 138.05 211.7 149.05 235.5 154.5 247.35 162.4 258.95 163.15 260.05 171.15 272.6 181.05 287.2 192.35 300.4 227.85 342 268 359.55 308.9 377.45 325.65 379.75 331.35 380.5 336.25 379.7 343.6 378.5 344.1 378.45 354.05 377.95 368.45 369.35 384.3 359.9 388.4 349.65 392.4 339.65 393.65 330.05 394.65 321.9 393.3 319.45 392.25 317.45 389 315.45 384.35 312.85 381.2 311.1 343.7 290.85 338.3 288.7 Z"></path> </svg> </div><div id="emailBtn" style="top: 10px; right: 214px; width: 14px; height: 14px; position: absolute; display: none; opacity: 0; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 483.3 483.3"><g fill="#fff"><path d="M424.3 57.75H59.1c-32.6 0-59.1 26.5-59.1 59.1v249.6c0 32.6 26.5 59.1 59.1 59.1h365.1c32.6 0 59.1-26.5 59.1-59.1v-249.5c.1-32.6-26.4-59.2-59-59.2zm32.1 308.7c0 17.7-14.4 32.1-32.1 32.1H59.1c-17.7 0-32.1-14.4-32.1-32.1v-249.5c0-17.7 14.4-32.1 32.1-32.1h365.1c17.7 0 32.1 14.4 32.1 32.1v249.5h.1z"></path><path d="M304.8 238.55l118.2-106c5.5-5 6-13.5 1-19.1-5-5.5-13.5-6-19.1-1l-163 146.3-31.8-28.4c-.1-.1-.2-.2-.2-.3-.7-.7-1.4-1.3-2.2-1.9L78.3 112.35c-5.6-5-14.1-4.5-19.1 1.1-5 5.6-4.5 14.1 1.1 19.1l119.6 106.9-119.1 111.5c-5.4 5.1-5.7 13.6-.6 19.1 2.7 2.8 6.3 4.3 9.9 4.3 3.3 0 6.6-1.2 9.2-3.6l120.9-113.1 32.8 29.3c2.6 2.3 5.8 3.4 9 3.4s6.5-1.2 9-3.5l33.7-30.2 120.2 114.2c2.6 2.5 6 3.7 9.3 3.7 3.6 0 7.1-1.4 9.8-4.2 5.1-5.4 4.9-14-.5-19.1l-118.7-112.7z"></path></g></svg></div></div></div><div id="liveIco" style="position: absolute; z-index: 11; width: 31px; height: 12px; bottom: 9px; left: 9px; display: none; cursor: pointer; pointer-events: auto; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); visibility: visible;"><svg id="Layer_1" data-name="Layer 1" xmlns="http://www.w3.org/2000/svg" viewBox="0 0 101.43 34.65"><defs><style>.clsx-1{isolation:isolate;}.clsx-2{fill:#ff3e5f;}.clsx-3{fill:#fff;}</style></defs><g id="_Live_1" data-name=" Live 1"><g class="clsx-1"><path class="clsx-2" d="M486.72,381l-.94,4.81H474.33l6.73-34.65h11.45l-.94,4.81h-6.12l-4.86,25Z" transform="translate(-474.33 -351.15)"></path></g><g class="clsx-1"><path class="clsx-3" d="M495,357.71h5.44l-3.09,15.89h8.47l-1,4.92H490.92Z" transform="translate(-474.33 -351.15)"></path><path class="clsx-3" d="M511.32,359.73h5.56l-3.65,18.8h-5.56Zm.51-4.74a4,4,0,0,1,3.79-3.17,2.52,2.52,0,0,1,2.55,3.17,4,4,0,0,1-3.78,3.13A2.53,2.53,0,0,1,511.83,355Z" transform="translate(-474.33 -351.15)"></path><path class="clsx-3" d="M518.82,359.73h5.63l2.6,11.94,7.21-11.94h5.63l-11.71,18.8h-5Z" transform="translate(-474.33 -351.15)"></path><path class="clsx-3" d="M537.81,369.09a12.07,12.07,0,0,1,12.09-9.85c5.82,0,8.83,4.25,7.8,9.55,0,0-.18.93-.42,1.79H543.08c-.32,2.24,1.33,3.69,4.28,3.69a7.41,7.41,0,0,0,4.94-1.83l2.64,3.1a12.91,12.91,0,0,1-8.76,3.43c-6.12,0-9.55-3.84-8.4-9.73ZM552.47,367c.33-1.72-1.22-3.13-3.46-3.13-2.42,0-4.67,1.38-5.16,3.13Z" transform="translate(-474.33 -351.15)"></path></g><path class="clsx-2" d="M564.63,381l4.86-25h-6.12l.94-4.81h11.45L569,385.8H557.58l.94-4.81Z" transform="translate(-474.33 -351.15)"></path></g></svg></div><div id="primisVastIconsContainer" style="user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"></div><div id="autoSkipContentContainer" style="width: 70px; height: 44px; position: absolute; right: 3px; top: 50%; transform: translateY(-50%); cursor: pointer; display: none; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="autoSkipContentNext" style="width: 67px; height: 20px; position: absolute; top: 0px; cursor: pointer; display: block; background-color: rgb(0, 0, 0); opacity: 0.75; border-radius: 5px; box-shadow: white 0px 0px 4px; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="autoSkipContentNextBg" style="width: 0px; position: absolute; cursor: pointer; display: block; background-color: rgb(255, 62, 95); border-radius: 5px; z-index: -1; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto; height: 20px;"></div><div id="autoSkipContentNextTxt" style="line-height: 20px; font-family: Roboto; font-size: 14px; font-weight: bold; text-align: center; color: rgb(255, 255, 255); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;">Next</div></div><div id="autoSkipContentStay" style="width: 67px; height: 20px; position: absolute; bottom: 0px; cursor: pointer; display: block; background-color: rgba(0, 0, 0, 0.75); border-radius: 5px; box-shadow: white 0px 0px 4px; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><div id="autoSkipContentStayTxt" style="line-height: 20px; font-family: Roboto; font-size: 14px; font-weight: bold; text-align: center; color: rgba(255, 255, 255, 0.8); cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;">Stay</div></div></div></div><div id="dcAdCover" style="position: absolute; top: 0px; left: 0px; z-index: 200; cursor: pointer; width: 100%; height: 100%; display: none; visibility: visible; pointer-events: auto;"></div></div><div id="adBreakDiv" style="width: 100%; height: 100%; position: absolute; display: none; z-index: 4; top: 0px; left: 0px; cursor: pointer;"><div id="adBreakPreloader" style="display: none;"></div></div></div><div id="lightBoxCloseBtn" style="position: absolute; z-index: 11; width: 24px; height: 24px; top: -28px; right: -28px; opacity: 1; display: none; cursor: pointer; user-select: none; -webkit-tap-highlight-color: rgba(0, 0, 0, 0); pointer-events: auto;"><svg xmlns="http://www.w3.org/2000/svg" width="100%" height="100%" style="width: 100%; height: 100%; pointer-events: none;" viewBox="0 0 475.2 475.2"><path fill="#fff" d="M405.6 69.6C360.7 24.7 301.1 0 237.6 0s-123.1 24.7-168 69.6S0 174.1 0 237.6s24.7 123.1 69.6 168 104.5 69.6 168 69.6 123.1-24.7 168-69.6 69.6-104.5 69.6-168-24.7-123.1-69.6-168zm-19.1 316.9c-39.8 39.8-92.7 61.7-148.9 61.7s-109.1-21.9-148.9-61.7c-82.1-82.1-82.1-215.7 0-297.8C128.5 48.9 181.4 27 237.6 27s109.1 21.9 148.9 61.7c82.1 82.1 82.1 215.7 0 297.8z" class="active-path" data-old_color="#A96D6D" data-original="#000000"></path><path fill="#fff" d="M342.3 132.9c-5.3-5.3-13.8-5.3-19.1 0l-85.6 85.6-85.6-85.6c-5.3-5.3-13.8-5.3-19.1 0-5.3 5.3-5.3 13.8 0 19.1l85.6 85.6-85.6 85.6c-5.3 5.3-5.3 13.8 0 19.1 2.6 2.6 6.1 4 9.5 4s6.9-1.3 9.5-4l85.6-85.6 85.6 85.6c2.6 2.6 6.1 4 9.5 4 3.5 0 6.9-1.3 9.5-4 5.3-5.3 5.3-13.8 0-19.1l-85.4-85.6 85.6-85.6c5.3-5.3 5.3-13.8 0-19.1z" class="active-path" data-old_color="#A96D6D" data-original="#000000"></path></svg></div><div id="pixelsDiv" style="width: 0px; height: 0px;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTI5JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MCZmqGE9MTplNwM2ODtzrD00MDAzrT0lMwUzqzyxX3Bup3NEo21unW49q3q3LzJcYzkyZ2F0ZXqurS5wo20zp3VvSWQ9q3q3LzJcYzkyZ2F0ZXqurS5wo20zZGVvqWqJozZipz1uqGyiow0znXNBpHA9MCZmZGg2PSZ1p2VlSXBBZGRlPTI0LwYkLwIlMS44MlZ1p2VlVUE9TW96nWkfYSUlRwUhMCUlMCUlOFqcozRiq3MyMwBOVCUlMDEjLwAyM0IyMwBXnW42NCUmQvUlMHt2NCUlOSUlMEFjpGkyV2VvS2y0JTJGNTM3LwM2JTIjJTI4S0uUTUjyMxMyMwBfnWgyJTIjR2Vwn28yMwxyMwBDnHJioWUyMxYkMwxhMC4jLwAyMwBTYWZupzxyMxY1MmphMmYyMwBFZGpyMxYkMwxhMC4jLwAzY3N1qWyxPTY2ZDBwNDY2YmMjNzIzY29hqGVhqEZcoGVJZD00NTMkNwA1Jz1yZGyuUGkurUkcp3RJZD0kNwE4MvZgZWRcYUkcp3RJZD00MwAlMCZwo250ZW50TWF0Y2uUrXByPSZcp0V4Y2k1ZGVGpz9gT3B0PTAzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA4MDM5OSZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD01ODZvMmpkMWI2ZTx4MwA4MwVzMwyyNzUlZWVyNGU2Mj==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTE2JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MCZmqGE9MTplNwM2ODtzrD00MDAzrT0lMwUzqzyxX3Bup3NEo21unW49q3q3LzJcYzkyZ2F0ZXqurS5wo20zp3VvSWQ9q3q3LzJcYzkyZ2F0ZXqurS5wo20zZGVvqWqJozZipz1uqGyiow0znXNBpHA9MCZmZGg2PSZ1p2VlSXBBZGRlPTI0LwYkLwIlMS44MlZ1p2VlVUE9TW96nWkfYSUlRwUhMCUlMCUlOFqcozRiq3MyMwBOVCUlMDEjLwAyM0IyMwBXnW42NCUmQvUlMHt2NCUlOSUlMEFjpGkyV2VvS2y0JTJGNTM3LwM2JTIjJTI4S0uUTUjyMxMyMwBfnWgyJTIjR2Vwn28yMwxyMwBDnHJioWUyMxYkMwxhMC4jLwAyMwBTYWZupzxyMxY1MmphMmYyMwBFZGpyMxYkMwxhMC4jLwAzY3N1qWyxPTY2ZDBwNDY2YmMjNzIzY29hqGVhqEZcoGVJZD00NTMkNwA2Jz1yZGyuUGkurUkcp3RJZD0kNwE4MvZgZWRcYUkcp3RJZD00MwAlMCZwo250ZW50TWF0Y2uUrXByPSZcp0V4Y2k1ZGVGpz9gT3B0PTEzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA4MDt4MSZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD1yYmM3MmQkN2Y4NGJvMWJuZwBuNWY1ZTAmNGY3ZWEkMQ==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTQlJaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw04MDMzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA4MwU2NlZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD1xNmFxOTUjY2ZzZGRuMwIlZzYmODM2MGFuYmYkNTU1MQ==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTI1JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw04MDMzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA4MwU3MCZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD0lMzU3YTYlZwA4ZDFxNGYkZDt4NDZuOTywYWE3Y2QjMt==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTQlJaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw03OTxzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA5MDU1NCZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD1vNTImYmE2NTMjOGMjMwxjNTywNTNuZDA2ODYlMGZuOA==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTI1JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw03OTxzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA5MDU1NvZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD00MGElZTRuN2ZwNGQ5MWZxN2ZxZGYkMmIjNmNwYWQ3ZQ==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTYzp2VlqzVlVGygZT0kNmI5MwtjMDA1JaZcZF9joGF5ZXJWZXI9Ml4kLwAzpm01ODA1NlZmqGE9MTplNwAkNwMzrD00MDAzrT0lMwUzoXN0YT0kNmI2MmY4OCZ2nWRsqzFmqFR5pGU9MlZ2nWRsqzyyq2FvnWkcqHyTqGF0ZT0kJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJaJ2ow0xr1ZQX1JWTy9NQUNST30zYXR0ZW1jqE11oHRcpGkcZXI9MwAzY29hqGVhqEZcoGVJZD0jJz1yZGyuUGkurUkcp3RJZD0jJz1yZGyuTGymqEyxPTAzpGkuY2VgZW50U3RlZWFgVHyjZT0mJzqxpHI9MCZaZHBlQ29hp2VhqD0znXNXZVBup3NHZHBlPTEzY2NjYT0jJzNwpGFDo25mZW50PSZwYaVmqGVlPTE3MwxlODAjOTE1MmtzqWyxPVNyn2yhZG9TUGkurWVlNwpkMzI4MDU0OTx5ZSZjqWJVpzj9nHR0pHMyM0EyMxYyMxZ3q3phYzyvoGVaYXRyq2F5LzNioSUlRzkyYXJhJTJGYzyvoGUgMTAkJTJGYWJiqXQgqGuyLWJcYzkyJTJGq2uyov13YXMgqGuyLWJcYzkyLXqlnXR0ZW4yMxYyM0Zzo3JgJTNETUpjQVYmJzZfo2F0U3RuqHVmPXRlqWUzZWyxp3A9nWykJaB4nWQ9NTUjZGU2MTp4OTU2YWUlOWJzYmuyMDYkMwI1YTx0MWE=" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTQlJaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw04MDAzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA5ODU1MvZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD0mY2VwMDA5ZzVuY2IjOGY2YwY3ODNwN2U5ZwYkOGU3ZQ==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTI1JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw04MDAzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDA5ODU1NvZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD0kYwM3NzElZWEmN2U4MmMlN2JuNzE1NTpjM2Q3MGVzZt==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTYzp2VlqzVlVGygZT0kNmI5MwtjMDA1JaZcZF9joGF5ZXJWZXI9Ml4kLwAzpm01ODA1NlZmqGE9MTplNTQ0MDEzrD00MDAzrT0lMwUzoXN0YT0kNmI2MmY4OCZ2nWRsqzFmqFR5pGU9MlZ2nWRsqzyyq2FvnWkcqHyTqGF0ZT0kJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJaJ2ow0xr1ZQX1JWTy9NQUNST30zYXR0ZW1jqE11oHRcpGkcZXI9MwAzY29hqGVhqEZcoGVJZD0jJz1yZGyuUGkurUkcp3RJZD0jJz1yZGyuTGymqEyxPTAzpGkuY2VgZW50U3RlZWFgVHyjZT0mJaNmpF9cqXM9MSZaZHBlPTAzZ2RjpxNioaNyoaQ9JzymV2VQYXNmR2Rjpw0kJzNwpGE9MCZwY3BuQ29hp2VhqD0zY2J1p3Rypw0kNmI5MwtjMTAlNTAlJaVcZD1TZWgcozRiU1BfYXyypwY3MTJvODA1NDx5OWUzpHVvVXJfPWu0qHBmJTNBJTJGJTJGq3q3LzJcYzkyZ2F0ZXqurS5wo20yMxZfZWFlovUlRzJcYzkyLTEjMSUlRzFvo3V0LXRbZS1vnWJfZSUlRaqbZW4gq2FmLXRbZS1vnWJfZS13pzy0qGVhJTJGJTNGZz9loSUmRE1HMEFWMlZzoG9uqFN0YXR1pm10paVyJzVcZHNjPWycpSZjrGyxPWVvZTx4NDt2YzUmYWVwMDY3NTI4Ymt2ZDQ2YTQ1MGMj" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTQ5JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJaZJoXBPpHBipaR1ozy0rU11oHRcpGkcZXI9NSZwo250ZW50RzyfZUyxPTAzoWVxnWFQoGF5TGymqEyxPTAzoWVxnWFMnXN0SWQ9MCZaZHBlPTAzZ2RjpxNioaNyoaQ9JzymV2VQYXNmR2Rjpw0kJzNwpGE9MCZwY3BuQ29hp2VhqD0zY2J1p3Rypw0kNmI5MwtjMTAlOTxmJaVcZD1TZWgcozRiU1BfYXyypwY3MTJvODA1NDx5OWUzpHVvVXJfPWu0qHBmJTNBJTJGJTJGq3q3LzJcYzkyZ2F0ZXqurS5wo20yMxZfZWFlovUlRzJcYzkyLTEjMSUlRzFvo3V0LXRbZS1vnWJfZSUlRaqbZW4gq2FmLXRbZS1vnWJfZS13pzy0qGVhJTJGJTNGZz9loSUmRE1HMEFWMlZzoG9uqFN0YXR1pm10paVyJzVcZHNjPWycpSZjrGyxPTt0NWQ5MWVwNTBwNmp1ZTU5MmU0ZzQ5ZTFvYmElZDQ4" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTE4JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MCZmqGE9MTplNwM2ODtzrD00MDAzrT0lMwUzqzyxX3Bup3NEo21unW49q3q3LzJcYzkyZ2F0ZXqurS5wo20zp3VvSWQ9q3q3LzJcYzkyZ2F0ZXqurS5wo20zZGVvqWqJozZipz1uqGyiow0znXNBpHA9MCZmZGg2PSZ1p2VlSXBBZGRlPTI0LwYkLwIlMS44MlZ1p2VlVUE9TW96nWkfYSUlRwUhMCUlMCUlOFqcozRiq3MyMwBOVCUlMDEjLwAyM0IyMwBXnW42NCUmQvUlMHt2NCUlOSUlMEFjpGkyV2VvS2y0JTJGNTM3LwM2JTIjJTI4S0uUTUjyMxMyMwBfnWgyJTIjR2Vwn28yMwxyMwBDnHJioWUyMxYkMwxhMC4jLwAyMwBTYWZupzxyMxY1MmphMmYyMwBFZGpyMxYkMwxhMC4jLwAzY3N1qWyxPTY2ZDBwNDY2YmMjNzIzY29hqGVhqEZcoGVJZD00NTMkNwA2Jz1yZGyuUGkurUkcp3RJZD0kNwE4MvZgZWRcYUkcp3RJZD00MwAlMCZwo250ZW50TWF0Y2uUrXByPSZcp0V4Y2k1ZGVGpz9gT3B0PTEzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDEjNwAlNCZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD0kMWM1ZwAmODAmMmt0MwU3ZWMkZWFxZGYlOGEkNzQjNj==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTI5JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MCZmqGE9MTplNwM2ODtzrD00MDAzrT0lMwUzqzyxX3Bup3NEo21unW49q3q3LzJcYzkyZ2F0ZXqurS5wo20zp3VvSWQ9q3q3LzJcYzkyZ2F0ZXqurS5wo20zZGVvqWqJozZipz1uqGyiow0znXNBpHA9MCZmZGg2PSZ1p2VlSXBBZGRlPTI0LwYkLwIlMS44MlZ1p2VlVUE9TW96nWkfYSUlRwUhMCUlMCUlOFqcozRiq3MyMwBOVCUlMDEjLwAyM0IyMwBXnW42NCUmQvUlMHt2NCUlOSUlMEFjpGkyV2VvS2y0JTJGNTM3LwM2JTIjJTI4S0uUTUjyMxMyMwBfnWgyJTIjR2Vwn28yMwxyMwBDnHJioWUyMxYkMwxhMC4jLwAyMwBTYWZupzxyMxY1MmphMmYyMwBFZGpyMxYkMwxhMC4jLwAzY3N1qWyxPTY2ZDBwNDY2YmMjNzIzY29hqGVhqEZcoGVJZD00NTMkNwA2Jz1yZGyuUGkurUkcp3RJZD0kNwE4MvZgZWRcYUkcp3RJZD00MwAlMCZwo250ZW50TWF0Y2uUrXByPSZcp0V4Y2k1ZGVGpz9gT3B0PTEzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDEjNwAlNSZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD05NGQ0MTBuN2M1ZTQmZDM3MDY2OTAmODNyY2M2ZDtlMt==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTE2JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MCZmqGE9MTplNwM2ODtzrD00MDAzrT0lMwUzqzyxX3Bup3NEo21unW49q3q3LzJcYzkyZ2F0ZXqurS5wo20zp3VvSWQ9q3q3LzJcYzkyZ2F0ZXqurS5wo20zZGVvqWqJozZipz1uqGyiow0znXNBpHA9MCZmZGg2PSZ1p2VlSXBBZGRlPTI0LwYkLwIlMS44MlZ1p2VlVUE9TW96nWkfYSUlRwUhMCUlMCUlOFqcozRiq3MyMwBOVCUlMDEjLwAyM0IyMwBXnW42NCUmQvUlMHt2NCUlOSUlMEFjpGkyV2VvS2y0JTJGNTM3LwM2JTIjJTI4S0uUTUjyMxMyMwBfnWgyJTIjR2Vwn28yMwxyMwBDnHJioWUyMxYkMwxhMC4jLwAyMwBTYWZupzxyMxY1MmphMmYyMwBFZGpyMxYkMwxhMC4jLwAzY3N1qWyxPTY2ZDBwNDY2YmMjNzIzY29hqGVhqEZcoGVJZD00NTMkNwA4Jz1yZGyuUGkurUkcp3RJZD0kNwE4MvZgZWRcYUkcp3RJZD00MwAlMCZwo250ZW50TWF0Y2uUrXByPSZcp0V4Y2k1ZGVGpz9gT3B0PTEzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDEjNwQ5NlZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD1uODE4ZwRuMzUjNGM2ZGNwYWE2NTFyYmE4MwM1Y2Q4ZA==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTQlJaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw04MDEzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDEjNwU1OSZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD0jYwQlMmBvY2IlYTQ2OTyuYmpkNDx0MwI1NzI1ZWUlOQ==" height="0px" width="0px" style="display: none !important;"><img src="https://live.primis.tech/live/liveView.php?hash=ozcmPTEznXRiPTEzqzyxX2V2ZW50PTI1JaNypaZypyRcoWU9MTplOTI4MDAjNSZ2nWRspGkurWVlVzVlPTMhMS4jJaM9MTE3Nwp3JaN0YT0jJat9NDAjJax9MwplJaZcZF9jYXNmRG9gYWyhPXq3ql5vnWJfZWquqGV3YXxhY29gJaN1YxyxPXq3ql5vnWJfZWquqGV3YXxhY29gJzRyYaVaSW5zo3JgYXRco249JzymQXBjPTAzp2Reqw0zqXNypxyjQWRxpw0lNC42MS4lMwEhODMzqXNypyVBPU1irzyfoGEyMxY1LwAyMwAyMwuXnW5xo3qmJTIjTyQyMwAkMC4jJTNCJTIjV2yhNwQyM0IyMwB4NwQyMwxyMwBBpHBfZVqyYxgcqCUlRwUmNl4mNvUlMCUlOEgIVE1MJTJDJTIjoGyeZSUlMEqyY2giJTI5JTIjQ2ulo21yJTJGMTI5LwAhMC4jJTIjU2FzYXJcJTJGNTM3LwM2JTIjRWRaJTJGMTI5LwAhMC4jJzNmqXVcZD02NzQjYmQ2NzMmMDZvJzNioaRyoaRGnWkySWQ9MCZgZWRcYVBfYXyMnXN0SWQ9MCZgZWRcYUkcp3RJZD0jJzR1pw04MDEzZ2Rjpw0jJzqxpHJDo25mZW50PSZcp1qyUGFmp0qxpHI9MSZwY3BuPTAzY2NjYUNioaNyoaQ9JzNvqXN0ZXI9MTplOTI4MDEjNwU2MSZ1nWQ9U2VenW5xo1NQoGF5ZXI2NmElYwtjNTQ5OTyyJaB1YyVloD1bqHRjplUmQSUlRvUlRaq3ql5vnWJfZWquqGV3YXxhY29gJTJGoGVupz4yMxZvnWJfZS0kMDEyMxZuYz91qC10nGUgYzyvoGUyMxZ3nGVhLXqupl10nGUgYzyvoGUgq3JcqHRyovUlRvUmRzZipz0yM0RNRmBBVwMzZzkiYXRTqGF0qXM9qHJ1ZSZynWRmpD1cnXEzpHucZD00MmE3NmRyODRuYmE2ODp3OTuxZzImODE1NTE4NWJvNj==" height="0px" width="0px" style="display: none !important;"></div></div></div></div></div></div></div><script src="https://live.primis.tech/live/liveView.php?s=117677&amp;cbuster=1729280007&amp;pubUrl=https://www.biblegateway.com/learn/bible-101/about-the-bible/when-was-the-bible-written/?form=MG0AV3&amp;x=400&amp;y=272&amp;playerApiId=freestarPrimisPlayer&amp;schain=1.0,1!freestar.com,1629,1" async="" type="text/javascript" id="freestarPrimisScript"></script></div>
 </div>
</div>
<style>
.td-fix-index {
  transform: initial !important;
  -webkit-transform: initial !important;
}
</style></div>



<p>But first, itâ€™s important to understand that the process of dating books of the Bible always involves a bit (or a lot) of guesswork.&nbsp;</p>



<h2 id="challenges-with-establishing-dates-in-the-bible" class="wp-block-heading">Challenges With Establishing Dates in the Bible&nbsp;</h2>



<p>There are several reasons why itâ€™s difficult to establish precise dates for when the Bible was written â€” including the lack of a consistent calendar, different understandings of history, and the common practice of oral transmission in the ancient world.&nbsp;&nbsp;</p>



<p>These problems diminish the closer you get to the modern age, but even in the New Testament era it can be difficult to lock down precise dating.&nbsp;</p>



<h3 id="no-consistent-calendar" class="wp-block-heading">No Consistent Calendar&nbsp;</h3>



<p>Before the adoption of the Gregorian calendar in 1582, there was no uniform dating system worldwide. Various cultures had different systems â€” some solar, some lunar, some based on rulersâ€™ reigns. Aligning these systems becomes increasingly challenging the further back you look.&nbsp;</p>



<p>To get past this challenge, historians meticulously reconstruct dates by cross-referencing rulersâ€™ reigns, celestial positions, other information given by ancient writers, and, more recently, carbon dating.&nbsp;&nbsp;</p>



<p>Still, dating precision varies, and can sometimes only be given in ranges of decades or even centuries.&nbsp;</p>



<h3 id="different-understanding-of-history" class="wp-block-heading">Different Understanding of History&nbsp;</h3>



<p>Another problem with dating ancient events is that the Bibleâ€™s authors had a different purpose for recording history than modern historians. While today we focus on establishing timelines and causality, ancient writers were more concerned with the eventsâ€™ <em>meanings </em>rather than their chronological order.&nbsp;&nbsp;</p>



<p>You can see this especially in the books of the Pentateuch (Genesis through Deuteronomy). Few of those stories provide any attempt at dating whatsoever. What matters is the way God appears in and uses history to accomplish his plan for his people.&nbsp;</p>



<h3 id="oral-transmission" class="wp-block-heading">Oral Transmission&nbsp;</h3>



<p>For many of us today, itâ€™s hard to imagine memorizing our email passwords, never mind the entire Bible. But in the ancient world, it was relatively common.&nbsp;</p>



<p>Paper in the ancient world â€” typically made painstakingly from cured animal skins â€” was very hard to come by, and literacy was rare. As a result, many works were transmitted orally for years, decades, or even centuries before they were written down.&nbsp;&nbsp;</p>



<p>This was especially true for popular stories, prophecies, Psalms and Proverbs, eyewitness accounts like the Gospels, and other genres that people might have shared widely before a scribe was able to commit them to paper.&nbsp;</p>



<h3 id="new-discoveries" class="wp-block-heading">New Discoveries&nbsp;</h3>



<p>Sometimes new academic or archaeological discoveries â€” ancient manuscripts, buried buildings, language evolution over time â€” reveal clues about when a Biblical work was composed. This can help determine more precise dating.&nbsp;&nbsp;</p>



<p>Today, scholars can date the reigns and activities of the <a href="https://www.biblegateway.com/learn/bible-101/kings-prophets-israel-judah/" target="_blank" rel="noreferrer noopener">kings of the books of Samuel, Kings, and Chronicles</a> with confidence thanks to the wealth of documents and material culture from that period. Even if that doesnâ€™t tell us exactly when the books <em>about</em> those kings were written, it can help give us an idea of the time frame.&nbsp;&nbsp;</p>



<p>Many of the prophets can be dated during that time as well, though others remain mysterious (Joel is the most notoriously difficult).&nbsp;</p>



<h2 id="dates-of-the-old-testament-books" class="wp-block-heading">Dates of the Old Testament Books&nbsp;</h2>



<p>With those caveats out of the way, here are our best guesses for when each book of the Bible was written, starting with the Old Testament.&nbsp;</p>



<h3 id="pentateuch-torah" class="wp-block-heading">Pentateuch (Torah)&nbsp;</h3>



<ul>
<li><strong>Genesis, Exodus, Leviticus, Numbers, Deuteronomy: </strong>1446-1406 BC; may have been mostly oral tradition until much later&nbsp;</li>
</ul>



<h3 id="historical-books" class="wp-block-heading">Historical Books&nbsp;</h3>



<ul>
<li><strong>Joshua: </strong>1400-1370 BC; possibly oral tradition recorded during the reign of Josiah (late 600s BC)&nbsp;</li>



<li><strong>Judges: </strong>1045-1000 BC; possibly oral tradition recorded during the reign of Josiah (late 600s BC)&nbsp;</li>



<li><strong>Ruth: </strong>1011-931 BC (likely during reign of David)&nbsp;</li>



<li><strong>1 &amp; 2 Samuel: </strong>930-722 BC; may have been mostly oral tradition compiled during the reign of Josiah (late 600s BC)&nbsp;</li>



<li><strong>1 &amp; 2 Kings: </strong>560-540 BC&nbsp;</li>



<li><strong>1 &amp; 2 Chronicles: </strong>450-425 BC&nbsp;</li>



<li><strong>Ezra: </strong>440-430 BC&nbsp;</li>



<li><strong>Nehemiah: </strong>430-400 BC&nbsp;</li>



<li><strong>Esther: </strong>Probably around 400 BC&nbsp;</li>
</ul>



<h3 id="wisdom-literature" class="wp-block-heading">Wisdom Literature&nbsp;</h3>



<ul>
<li><strong>Job: </strong>Disputed; dates range from second millennium to fifth century BC&nbsp;</li>



<li><strong>Psalms: </strong>Various times from 1400-500 BC&nbsp;</li>



<li><strong>Proverbs:</strong> 950-700 BC&nbsp;</li>



<li><strong>Ecclesiastes: </strong>Around 935 BC&nbsp;</li>



<li><strong>Song of Solomon: </strong>960-931 BC&nbsp;</li>
</ul>



<h3 id="prophets" class="wp-block-heading">Prophets&nbsp;</h3>



<ul>
<li><strong>Isaiah:</strong> 700-681 BC; oral traditions may have been compiled much later, with additional content as late as early 500s BC&nbsp;</li>



<li><strong>Jeremiah:</strong> 626-585 BC&nbsp;</li>



<li><strong>Lamentations:</strong> ca. 586 BC&nbsp;</li>



<li><strong>Ezekiel:</strong> 593-571 BC&nbsp;</li>



<li><strong>Daniel:</strong> Beginning 530 BC; some parts may be much later&nbsp;</li>



<li><strong>Hosea:</strong> 750-715 BC&nbsp;</li>



<li><strong>Joel: </strong>Disputed; likely either 835-796 BC or 500-450 BC&nbsp;</li>



<li><strong>Amos:</strong> 760-750 BC&nbsp;</li>



<li><strong>Obadiah:</strong> Disputed; 850-450 BC, though most likely the early exilic period (580-560 BC)&nbsp;</li>



<li><strong>Jonah: </strong>Probably 500s BC, though it takes place much earlier (790-760 BC)&nbsp;</li>



<li><strong>Micah:</strong> 735-700 BC&nbsp;</li>



<li><strong>Nahum: </strong>663-612 BC&nbsp;</li>



<li><strong>Habakkuk:</strong> 612-589 BC&nbsp;</li>



<li><strong>Zephaniah:</strong> 640-609 BC&nbsp;</li>



<li><strong>Haggai:</strong> Around 520 BC&nbsp;</li>



<li><strong>Zechariah:</strong> 520-480 BC&nbsp;</li>



<li><strong>Malachi:</strong> 440-430 BC&nbsp;</li>
</ul>



<h2 id="dates-of-the-apocrypha" class="wp-block-heading">Dates of the Apocrypha&nbsp;</h2>



<p>In general, the Deuterocanonical or Apocryphal books were <a href="https://www.biblegateway.com/learn/bible-101/original-language-of-the-bible/" target="_blank" rel="noreferrer noopener">written in Greek rather than Hebrew</a> and date from the Intertestamental Period â€” roughly 300-100 BC or a little later.&nbsp;</p>



<h2 id="dates-of-the-new-testament-books" class="wp-block-heading">Dates of the New Testament Books&nbsp;</h2>



<p>The books of the New Testament are much easier to date than the Old Testament. They were obviously all written after the death and resurrection of Jesus in 33 AD, and were mostly in their completed forms by 100 AD, when Christian writers began referencing and quoting from them.&nbsp;&nbsp;</p>



<p>Still, some of the same problems persist from the Old Testament period, and there is dispute about the exact dating of the Gospels (most of which were oral testimony written down after the fact) and most of the epistles. Unlike the Old Testament, though, they are disputes of a few years or, at the most, decades â€” rather than centuries.&nbsp;</p>



<h3 id="gospels-acts" class="wp-block-heading">Gospels &amp; Acts&nbsp;</h3>



<ul>
<li><strong>Matthew:</strong> After 70 AD, preserving oral accounts from the previous generation&nbsp;</li>



<li><strong>Mark: </strong>64-70 AD, preserving oral accounts from the previous generation&nbsp;</li>



<li><strong>Luke:</strong> Debated; could be any time between 62-90 AD, preserving earlier oral accounts&nbsp;</li>



<li><strong>John: </strong>90-110 AD, preserving earlier oral accounts&nbsp;</li>



<li><strong>Acts: </strong>62-90 AD (by Luke)&nbsp;</li>
</ul>



<h3 id="pauline-epistles" class="wp-block-heading">Pauline Epistles&nbsp;</h3>



<ul>
<li><strong>Romans: </strong>56-57 AD&nbsp;</li>



<li><strong>1 Corinthians:</strong> 53-54 AD&nbsp;</li>



<li><strong>2 Corinthians: </strong>55-56 AD&nbsp;</li>



<li><strong>Galatians: </strong>50-56 AD&nbsp;</li>



<li><strong>Ephesians:</strong> 60-62 AD&nbsp;</li>



<li><strong>Philippians:</strong> 54-62 AD, depending on whether it was sent from Ephesus, Caesarea, or Rome&nbsp;</li>



<li><strong>Colossians:</strong> 57-62 AD, depending on whether it was sent from Caesarea or Rome&nbsp;</li>



<li><strong>1 Thessalonians:</strong> 50-51 AD&nbsp;</li>



<li><strong>2 Thessalonians:</strong> 51-52 AD&nbsp;</li>



<li><strong>1 Timothy:</strong> 62-64 AD (or 90-110 AD if pseudonymous)&nbsp;</li>



<li><strong>2 Timothy:</strong> 64-67 AD (or 90-110 AD if pseudonymous)&nbsp;</li>



<li><strong>Titus: </strong>62-64 AD (or 90-110 AD if pseudonymous)&nbsp;</li>



<li><strong>Philemon:</strong> 54-62 AD, depending on whether it was sent from Ephesus, Caesarea, or Rome&nbsp;</li>
</ul>



<h3 id="general-epistles-revelation" class="wp-block-heading">General Epistles &amp; Revelation&nbsp;</h3>



<ul>
<li><strong>Hebrews: </strong>60-95 AD, but probably late 60s&nbsp;</li>



<li><strong>James:</strong> 45-62 AD&nbsp;</li>



<li><strong>1 Peter:</strong> 60-65 AD (or 85-90 AD if pseudonymous)&nbsp;</li>



<li><strong>2 Peter: </strong>65-68 AD (or 80-90 AD if pseudonymous)&nbsp;</li>



<li><strong>1 John:</strong> 85-100 AD&nbsp;</li>



<li><strong>2 John:</strong> 85-100 AD&nbsp;</li>



<li><strong>3 John:</strong> 85-100 AD&nbsp;</li>



<li><strong>Jude:</strong> 65-80 AD&nbsp;</li>



<li><strong>Revelation:</strong> Either 64-65 AD (during persecutions by Nero) or 95-96 AD (during persecutions by Domitian)&nbsp;</li>
</ul>



<h2 id="conclusion-when-did-each-book-take-place" class="wp-block-heading">Conclusion: When Did Each Book Take Place?&nbsp;</h2>



<p>Understanding even roughly when the books of the Bible were written helps to contextualize not only their relationships to each other, but also to the wider cultures of Israel, the Levant, and the Middle East at the time of their composition.&nbsp;</p>



<p>You also might notice that they werenâ€™t always written (or at least not written <em>down</em>) exactly when they took place â€” especially with books like Jonah and Esther that take place decades or even centuries before they were finalized.&nbsp;</p>



<p>If youâ€™d like to read the books in the order they take place, look no further than our <a href="https://www.biblegateway.com/learn/bible-101/books-of-the-bible-in-order/" target="_blank" rel="noreferrer noopener">chronological ordering of the Bible</a> â€” and learn more about the circumstances of their composition by digging into <a href="https://www.biblegateway.com/learn/bible-101/who-wrote-the-bible/" target="_blank" rel="noreferrer noopener">who wrote each book</a>.&nbsp;</p>



<p><strong>Want to get better context behind each of the dates in this article? Try </strong><a href="https://www.biblegateway.com/plus/?utm_source=bg&amp;utm_medium=bglearn" target="_blank" rel="noreferrer noopener"><strong>Bible Gateway Plus</strong></a><strong> free for 14 days and get access to some of the same resources I used to research this very article!</strong>&nbsp;</p>

<!-- MOLONGUI AUTHORSHIP PLUGIN 4.9.3 -->
<!-- https://www.molongui.com/wordpress-plugin-post-authors -->

<div class="molongui-clearfix"></div><div id="mab-5934319882" class="m-a-box " data-box-layout="slim" data-box-position="below" data-multiauthor="false" data-author-id="3" data-author-type="user" data-author-archived=""><div class="m-a-box-container" style="position: relative;" min-width="600px"><div class="m-a-box-tab m-a-box-content m-a-box-profile" data-profile-layout="layout-1" data-author-ref="user-3" itemscope="" itemid="https://www.biblegateway.com/learn/author/jacob-edson/" itemtype="https://schema.org/Person"><div class="m-a-box-content-top"></div><div class="m-a-box-content-middle"><div class="m-a-box-item m-a-box-avatar" data-source="local"><a class="m-a-box-avatar-url" href="https://www.biblegateway.com/learn/author/jacob-edson/"><img decoding="async" width="150" height="150" src="https://i0.wp.com/staticblog.bgcdn.com/learn/wp-content/uploads/2024/08/16165603/jacob-edson-scaled-e1723841809840-150x150.jpg?resize=150%2C150&amp;ssl=1" class="attachment-150x150 size-150x150" alt="Jacob Edson headshot" itemprop="image" data-recalc-dims="1"></a></div><div class="m-a-box-item m-a-box-data"><div class="m-a-box-name m-a-box-title"><h5 id="jacob-edson" itemprop="name"><a class="m-a-box-name-url " href="https://www.biblegateway.com/learn/author/jacob-edson/" itemprop="url"> Jacob Edson</a></h5></div><div class="m-a-box-item m-a-box-meta"></div><div class="m-a-box-bio" itemprop="description"><p>Jacob is Editorial Director of Bible Gateway. He holds a Master of Theological Studies in Early Christian Thought from Harvard Divinity School, and a Bachelor of Arts in Religious History from Memorial University of Newfoundland, though with most of his coursework from the University of Hawai'i at Manoa. His work has appeared in Ekstasis and in Geez Magazine's "Embracing Darkness" Advent devotional.</p></div><div class="m-a-box-related" data-related-layout="layout-1"><div class="m-a-box-item m-a-box-related-entries" style="display:none;"><ul><li><div class="m-a-box-related-entry" itemscope="" itemtype="http://schema.org/CreativeWork"><div class="molongui-display-none" itemprop="author" itemscope="" itemtype="http://schema.org/Person"><div itemprop="name">Jacob Edson</div><div itemprop="url">https://www.biblegateway.com/learn/author/jacob-edson/</div></div><div class="m-a-box-related-entry-title"><i class="m-a-icon-doc"></i><a class="molongui-remove-underline" href="https://www.biblegateway.com/learn/bible-101/about-the-bible/septuagint/" itemprop="url"><span itemprop="headline">What Is the Septuagint? Discover the Worldâ€™s First Bible Translation</span></a></div></div></li><li><div class="m-a-box-related-entry" itemscope="" itemtype="http://schema.org/CreativeWork"><div class="molongui-display-none" itemprop="author" itemscope="" itemtype="http://schema.org/Person"><div itemprop="name">Jacob Edson</div><div itemprop="url">https://www.biblegateway.com/learn/author/jacob-edson/</div></div><div class="m-a-box-related-entry-title"><i class="m-a-icon-doc"></i><a class="molongui-remove-underline" href="https://www.biblegateway.com/learn/bible-101/about-the-bible/ten-commandments/" itemprop="url"><span itemprop="headline">What Are the 10 Commandments? Where to Find Them in the Bible</span></a></div></div></li><li><div class="m-a-box-related-entry" itemscope="" itemtype="http://schema.org/CreativeWork"><div class="molongui-display-none" itemprop="author" itemscope="" itemtype="http://schema.org/Person"><div itemprop="name">Jacob Edson</div><div itemprop="url">https://www.biblegateway.com/learn/author/jacob-edson/</div></div><div class="m-a-box-related-entry-title"><i class="m-a-icon-doc"></i><a class="molongui-remove-underline" href="https://www.biblegateway.com/learn/bible-101/about-the-bible/lost-books-bible/" itemprop="url"><span itemprop="headline">Are There Books Missing from the Bible? The Truth About the 'Lost Books'</span></a></div></div></li><li><div class="m-a-box-related-entry" itemscope="" itemtype="http://schema.org/CreativeWork"><div class="molongui-display-none" itemprop="author" itemscope="" itemtype="http://schema.org/Person"><div itemprop="name">Jacob Edson</div><div itemprop="url">https://www.biblegateway.com/learn/author/jacob-edson/</div></div><div class="m-a-box-related-entry-title"><i class="m-a-icon-doc"></i><a class="molongui-remove-underline" href="https://www.biblegateway.com/learn/bible-101/best-study-bibles/" itemprop="url"><span itemprop="headline">33 of the Best Study Bibles for Every Type of Reader [2024]</span></a></div></div></li></ul></div></div></div></div><div class="m-a-box-content-bottom"></div></div><div dir="ltr" class="resize-sensor" style="pointer-events: none; position: absolute; inset: 0px; overflow: hidden; z-index: -1; visibility: hidden; max-width: 100%;"><div class="resize-sensor-expand" style="pointer-events: none; position: absolute; inset: 0px; overflow: hidden; z-index: -1; visibility: hidden; max-width: 100%;"><div style="position: absolute; left: 0px; top: 0px; transition: all; width: 842px; height: 209px;"></div></div><div class="resize-sensor-shrink" style="pointer-events: none; position: absolute; inset: 0px; overflow: hidden; z-index: -1; visibility: hidden; max-width: 100%;"><div style="position: absolute; left: 0px; top: 0px; transition: all; width: 200%; height: 200%;"></div></div></div></div></div></div>'''
markdown = md(html)
pyperclip.copy(markdown)
```


Diagram Explanation: Meritocracy, Competency Crisis, and Overproduction of Elites
=================================================================================


To conceptualize the relationship between Meritocracy, Competency Crisis, and Overproduction of Elites, we can visualize these concepts as nodes on a graph, with connecting lines representing their interrelationships. Below is a description of how these nodes might be connected:
Nodes and Connections
---------------------


1. **Meritocracy Node**
	* **Connection to Competency Crisis**: Meritocracy is intended to reward individuals based on their abilities and achievements. However, the system often leads to the promotion of individuals who may not be truly competent, contributing to a competency crisis. This occurs when meritocratic systems fail to accurately assess and promote true competence, leading to the erosion of institutional effectiveness [1](https://insights.som.yale.edu/insights/how-meritocracy-worsens-inequality-and-makes-even-the-rich-miserable#:~:text=How%20Meritocracy%20Worsens%20Inequality%E2%80%94and,Makes%20Even%20the%20Rich).
	* **Connection to Overproduction of Elites**: Meritocracy contributes to the overproduction of elites by creating a cycle where educational and career opportunities are concentrated among a select group. This results in an oversupply of highly educated individuals competing for limited elite positions, leading to underemployment and social instability [2](https://insights.som.yale.edu/insights/how-meritocracy-worsens-inequality-and-makes-even-the-rich-miserable#:~:text=The%20avenues%20that%20once,American%20elite%20are%20narrowing).
2. **Competency Crisis Node**
	* **Connection to Meritocracy**: The competency crisis is exacerbated by meritocratic systems that prioritize credentials and perceived merit over actual competence. This leads to the systematic promotion of unqualified individuals, further eroding institutional competency [1](https://insights.som.yale.edu/insights/how-meritocracy-worsens-inequality-and-makes-even-the-rich-miserable#:~:text=How%20Meritocracy%20Worsens%20Inequality%E2%80%94and,Makes%20Even%20the%20Rich).
	* **Connection to Overproduction of Elites**: The competency crisis is linked to the overproduction of elites as the abundance of highly educated individuals does not necessarily translate into competent leadership or effective institutional performance. This mismatch contributes to systemic failures and social unrest [3](https://insights.som.yale.edu/insights/how-meritocracy-worsens-inequality-and-makes-even-the-rich-miserable#:~:text=In%20the%20middle%20of,embraced%20the%20idea%20of).
3. **Overproduction of Elites Node**
	* **Connection to Meritocracy**: The overproduction of elites is driven by meritocratic systems that emphasize educational attainment and elite status. As more individuals achieve high levels of education, the competition for elite positions intensifies, leading to underemployment and dissatisfaction among the educated class [1](https://insights.som.yale.edu/insights/how-meritocracy-worsens-inequality-and-makes-even-the-rich-miserable#:~:text=How%20Meritocracy%20Worsens%20Inequality%E2%80%94and,Makes%20Even%20the%20Rich).
	* **Connection to Competency Crisis**: The surplus of elites can lead to a competency crisis when the focus on credentials overshadows the need for genuine competence. This results in a disconnect between the qualifications of individuals and their ability to effectively contribute to institutional success [3](https://insights.som.yale.edu/insights/how-meritocracy-worsens-inequality-and-makes-even-the-rich-miserable#:~:text=In%20the%20middle%20of,embraced%20the%20idea%20of).


Diagram Summary
---------------


In summary, the diagram would feature three interconnected nodes: Meritocracy, Competency Crisis, and Overproduction of Elites. Each node is linked to the others, illustrating the cyclical and interdependent nature of these phenomena. Meritocracy feeds into both the competency crisis and the overproduction of elites, while the competency crisis and overproduction of elites reinforce each other, creating a complex web of challenges that modern societies must navigate.

```python

import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define nodes
nodes = ["Meritocracy", "Competency\nCrisis", "Overproduction\nof Elites"]

# Add nodes to the graph
G.add_nodes_from(nodes)

# Define edges (causal relationships)
edges = [
    ("Meritocracy", "Competency\nCrisis"),
    ("Meritocracy", "Overproduction\nof Elites"),
    ("Competency\nCrisis", "Overproduction\nof Elites"),
    ("Overproduction\nof Elites", "Competency\nCrisis"),
]
edge_labels = {("Meritocracy", "Competency\nCrisis"): "+", 
               ("Meritocracy", "Overproduction\nof Elites"): "+", 
               ("Competency\nCrisis", "Overproduction\nof Elites"): "-", 
               ("Overproduction\nof Elites", "Competency\nCrisis"): "-"}

# Add edges to the graph
G.add_edges_from(edges)

# Define positions for nodes
pos = nx.spring_layout(G)
```

```python

# Create a figure and axis
# fig_height = 18/nu.twitter_aspect_ratio
fig_width = 18; fig_height = 6
fig, ax = plt.subplots(figsize=(fig_width, fig_height))

# Draw the nodes
nx.draw_networkx_nodes(G, pos, node_size=3000, node_color='lightblue', alpha=0.5, ax=ax)

# Draw the edges with curved lines and arrowheads
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=20, connectionstyle='arc3,rad=0.2', ax=ax)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold', ax=ax)

# Draw edge labels to indicate the nature of the relationships
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, ax=ax)

# Set limits to avoid clipping
# ax.set_xlim(-1.5, 1.5) 
# ax.set_ylim(-1.5, 1.5)

# Show the plot
ax.set_title("Causal Loop Diagram with Curved Edges and Arrowheads")
ax.axis('off')  # Hide the axes
plt.show()
```
