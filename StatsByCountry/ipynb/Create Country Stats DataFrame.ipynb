```python

union_set = set()
svg_dict = {}
for wikipedia_url in all_countries_df.wikipedia_url.tolist():
    page_title = wikipedia_url.split('/')[-1]
    try:
        page_obj = wikipedia.page(title=page_title)
        image_list = [img for img in page_obj.images if img.endswith('.svg')]
        svg_dict[page_title] = image_list
        for image_url in image_list:
            union_set.add(image_url)
    except Exception as e:
        print(str(e).strip())
    #print(['page_obj.{}'.format(fn) for fn in dir(page_obj) if not fn.startswith('_')])
    #break
```
