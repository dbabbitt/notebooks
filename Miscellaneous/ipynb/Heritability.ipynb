```python

import pandas as pd

heritable_str = '''Human_Trait,Average_Ratings,Genetic_Research
Eye Color,0.77,0.95
Height,0.67,0.8
Weight,0.4,0.7
Breast Cancer,0.53,0.1
Stomach Ulcers,0.29,0.7
Schizophrenia,0.43,0.5
Autism,0.42,0.7
Reading Disability,0.38,0.6
School Achievement,0.29,0.6
Verbal Ability,0.27,0.6
Remembering Faces,0.31,0.6
Spatial Ability,0.33,0.7
General Intelligence,0.41,0.5
Personality,0.38,0.4'''
heritable_df = pd.read_csv(pd.compat.StringIO(heritable_str), sep=',')
```

```python

from random import shuffle

questionairre_str = 'According to the latest genetic research, '
questionairre_str += 'which Percent Heritability goes with which Human Trait?'
df = heritable_df.sample(5).sort_index()
print(questionairre_str)
print()
i = ord('A') # 65
for row_index, row_series in df.iterrows():
    print('{}. {}%'.format(chr(i), round(row_series['Genetic_Research']*100, 1)))
    i += 1
print()
answer_list = df['Human_Trait'].tolist()
shuffle(answer_list)
i = ord('1') # 49
for answer in answer_list:
    print('{}. {}'.format(chr(i), answer))
    i += 1
```

```python

def f(row_series):
    
    return abs(row_series['Average_Ratings'] - row_series['Genetic_Research'])

heritable_df['diff'] = heritable_df.apply(f, axis=1)
```

```python

heritable_df.sort_values('diff', ascending=False)
```

```python

from random import shuffle

questionairre_str = 'BOSS LEVEL: According to the latest genetic research, '
questionairre_str += 'which Percent Heritability goes with which Human Trait?'
df = heritable_df.sort_values('diff', ascending=False).head(5).sort_index()
print(questionairre_str)
print()
question_list = df['Genetic_Research'].tolist()
question_list = sorted(question_list)
i = ord('A') # 65
for question in question_list:
    print('{}. {}%'.format(chr(i), float(question)*100))
    i += 1
print()
answer_list = df['Human_Trait'].tolist()
answer_list = sorted(answer_list)
i = ord('1') # 49
for answer in answer_list:
    print('{}. {}'.format(chr(i), answer))
    i += 1
```
