```python

import networkx as nx

tribe_digraph = nx.DiGraph()
tribe_digraph.add_nodes_from(tribe_df.T.to_dict())
```

```python

list(tribe_digraph.nodes(data=True))
```

```python

from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

tribe_list = tribe_df.index.tolist()
for row_index, row_series in pd.DataFrame(tribe_df['COMBATANTS']).iterrows():
    combatant_list = row_series['COMBATANTS'].split(', ')
    for combatant in combatant_list:
        if combatant != 'None':
            this_tribe = sorted(tribe_list, key=lambda tribe: similar(combatant, tribe.split(' (')[0]))[-1]
            tribe_digraph.add_edge(row_index, this_tribe)
```

```python

import matplotlib.pyplot as plt
%matplotlib inline

# Draw the graph using the default spring layout
plt.figure(figsize=(10,9))
nx.draw_networkx(tribe_digraph)
```

```python

# Draw the graph using the spring layout
plt.figure(figsize=(10,9))
spring_pos = nx.spring_layout(tribe_digraph)
nx.draw_networkx(tribe_digraph, spring_pos)
```

```python

# Export the graph into Gephi
file_path = SAVES_FOLDER+'/gexf/tribe_digraph.gexf'
nx.write_gexf(tribe_digraph, file_path)
```

```python

full_edge_list = list(tribe_digraph.edges())
double_edge_list = []
while len(full_edge_list) > 0:
    edge_tuple = full_edge_list.pop()
    doubled_tuple = (edge_tuple[1], edge_tuple[0])
    if doubled_tuple in full_edge_list:
        double_edge_list.append(edge_tuple)
        double_edge_list.append(doubled_tuple)
double_edge_list
```

```python

single_edge_list = []
full_edge_list = list(tribe_digraph.edges())
for edge_tuple in full_edge_list:
    if edge_tuple not in double_edge_list:
        single_edge_list.append(edge_tuple)
single_edge_list
```

```python

# Import from Gephi
import xmltodict
import numpy as np
import matplotlib.patches as mpatches

with open(file_path, 'r') as fid:
    tribal_xml = fid.read()
tribal_tree = xmltodict.parse(tribal_xml)

# Get the position dictionary for the GEXF saved from Gephi
node_dict_list = tribal_tree['gexf']['graph']['nodes']['node']
tribal_layout_dict = {node_dict['@id']: np.array([float(node_dict['viz:position']['@x']),
                                                  float(node_dict['viz:position']['@y'])]) for node_dict in node_dict_list}

# Draw the graph using the gephi layout
tribal_fig = plt.figure(figsize=(18, 9))
ax = tribal_fig.add_subplot(111)
plt.axis('off')
plt.xticks([], [])
plt.yticks([], [])
arrowstyle = mpatches.ArrowStyle.Simple(head_length=2., head_width=1.5, tail_width=0.05)

# Make the nodes blue so you can make the double edges red
node_collection = nx.draw_networkx_nodes(G=tribe_digraph, pos=tribal_layout_dict, alpha=0.5, node_color='b')

# Get the double-edged edge list in red
edge_collection = nx.draw_networkx_edges(G=tribe_digraph, pos=tribal_layout_dict, alpha=0.25,
                                         arrowstyle=arrowstyle, edgelist=single_edge_list)
edge_collection = nx.draw_networkx_edges(G=tribe_digraph, pos=tribal_layout_dict, alpha=0.25,
                                         arrowstyle=arrowstyle, edgelist=double_edge_list, edge_color='r')

labels_collection = nx.draw_networkx_labels(G=tribe_digraph, pos=tribal_layout_dict, font_size=14)
axes_subplot = tribal_fig.get_axes()[0]
xlim_tuple = axes_subplot.set_xlim(left=-1313, right=1271)
```
