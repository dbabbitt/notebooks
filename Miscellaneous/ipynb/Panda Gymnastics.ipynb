```python

# Don't run this if you can already see the image below
from PIL import Image
import requests
from io import BytesIO
from IPython.display import HTML
import base64

panda_url = 'https://secure.i.telegraph.co.uk/multimedia/archive/02152/panda-upside-down_2152868i.jpg'
panda_response = requests.get(panda_url)
panda_img = BytesIO(panda_response.content).getvalue()
base64_byte_str = base64.b64encode(panda_img).decode('ascii')
img_src = 'data:image/JPEG;base64,{}'.format(base64_byte_str)
img_tag = '<img src="{}" style="height:100%;" />'.format(img_src)
HTML(img_tag)
```
