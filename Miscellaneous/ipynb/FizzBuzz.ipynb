```python

import itertools

fizzes = itertools.cycle(['', '', 'fizz'])
buzzes = itertools.cycle(['', '', '', '', 'buzz'])
numbers = itertools.count(1)
fizz_buzzes = ((fizz + buzz) or str(n) for fizz, buzz, n in zip(fizzes, buzzes, numbers))
fb_list = [next(fizz_buzzes) for _ in range(100)]
for fb in fb_list:
    print(fb)
```
