```python

from PIL import Image
import os
import pandas as pd

screen_width = 1920
screen_height = 1080
screen_aspect_ratio = screen_width / screen_height
wallpaper_folder = r'C:\Users\Dave Babbitt\Pictures\Wallpaper'
rows_list = []
for sub_dir, dir_list, file_list in os.walk(wallpaper_folder):
    for src_file in file_list:
        src_path = os.path.join(sub_dir, src_file)
        image_file = Image.open(src_path)
        image_width, image_height = image_file.size
        row_dict = {}
        row_dict['src_file'] = src_file
        row_dict['src_path'] = src_path
        row_dict['image_width'] = image_width
        row_dict['image_height'] = image_height
        row_dict['image_format'] = image_file.format
        rows_list.append(row_dict)

column_list = ['src_file', 'src_path', 'image_width', 'image_height', 'image_format']
wallpaper_df = pd.DataFrame(rows_list, columns=column_list)
```

```python

wallpaper_df['image_width'] = wallpaper_df['image_width'].map(lambda x: int(x))
wallpaper_df['image_height'] = wallpaper_df['image_height'].map(lambda x: int(x))
wallpaper_df['aspect_ratio'] = wallpaper_df.apply(lambda row: row['image_width']/row['image_height'], axis=1)
wallpaper_df['image_size'] = wallpaper_df.apply(lambda row: row['image_width']*row['image_height'], axis=1)
wallpaper_df['aspect_distance'] = wallpaper_df['aspect_ratio'].map(lambda x: abs(screen_aspect_ratio-x))
column_list = ['src_file', 'image_format', 'image_width', 'image_height', 'aspect_ratio', 'image_size', 'aspect_distance']
wallpaper_df.sort_values('aspect_distance', ascending=False)[column_list].head(10)
```

```python

mask_series = (wallpaper_df['aspect_distance'] >= 1.0)
for row_index, row_series in wallpaper_df[mask_series].iterrows():
    file_path = row_series['src_path']
    os.remove(file_path)
```

```python

from bs4 import BeautifulSoup
import requests

aspect_ratios_path = '../data/html/aspect_ratios.html'
with open(aspect_ratios_path, 'r') as text_io_wrapper:
    aspect_ratios_str = text_io_wrapper.read()
aspect_ratios_soup = BeautifulSoup(aspect_ratios_str, 'lxml')
```

```python

import re

non_numeric_regex = re.compile(r'[^\d.]+')
aspect_ratio_dict = {}
for li_tag in aspect_ratios_soup.find_all('li'):
    for b_tag in li_tag.find_all('b'):
        aspect_ratio_str = b_tag.text
        try:
            aspect_ratio_float = float(non_numeric_regex.sub('', aspect_ratio_str.split(' ')[0].split(':')[0]))
        except:
            print(aspect_ratio_str)
    aspect_ratio_descr = li_tag.text[len(aspect_ratio_str)+2:]
    aspect_ratio_dict[aspect_ratio_float] = aspect_ratio_descr
```

```python

mask_series = wallpaper_df['aspect_ratio'].isin(list(aspect_ratio_dict.keys()))
for aspect_ratio in wallpaper_df[mask_series]['aspect_ratio'].unique():
    print()
    print(aspect_ratio)
    print(aspect_ratio_dict[aspect_ratio])
```
