```python

%run ../load_magic/storage.py
from datetime import datetime
import pandas as pd
import twarc
import json

secrets_path = '../data/json/secrets.json'
with open(secrets_path) as f:
    secrets_dict = json.loads(f.read())
consumer_key = secrets_dict['twitter_api_key']
consumer_secret = secrets_dict['twitter_api_key_secret']
access_token_key = secrets_dict['twitter_access_token']
access_token_secret = secrets_dict['twitter_access_token_secret']
tw2 = twarc.Twarc2(consumer_key, consumer_secret, access_token_key, access_token_secret)
s = Storage()
usernames_list = s.load_object('usernames_list')
userids_list = s.load_object('userids_list')
iso_8601_ms = '%Y-%m-%dT%H:%M:%S.%fZ'
rows_list = []
for username, userid in zip(usernames_list, userids_list):
    row_dict = {}
    row_dict['user_name'] = username
    row_dict['userid'] = username
    latest_datetime = pd.NaT
    search_recents_list = [t for t in tw2.search_recent(f'from:{username}')]
    if search_recents_list:
        latest_datetime = datetime.strptime(search_recents_list[0]['data'][0]['created_at'], iso_8601_ms)
    row_dict['latest_datetime'] = latest_datetime
    rows_list.append(row_dict)
followings_df = pd.DataFrame(rows_list)
s.store_objects(followings_df=followings_df)
```
