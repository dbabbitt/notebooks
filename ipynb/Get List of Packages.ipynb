```python

from pip._vendor import pkg_resources

dict = pkg_resources.WorkingSet().entry_keys
for key in dict:
    for name in dict[key]:
        req = pkg_resources.working_set.by_key[name].requires()
        if len(req):
            print()
            print('{} from {} requires:'.format(name, key))
            for r in req:
                print('{}'.format(r))
```
