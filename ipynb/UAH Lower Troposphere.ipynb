```python

import pandas as pd

data_url = "http://www.nsstc.uah.edu/data/msu/v6.0beta/tlt/uahncdc_lt_6.0beta5.txt"
CSV_PARAMS = dict(
    header=0,
    index_col=0,
    skipinitialspace=True,
    na_filter=False,
    skipfooter=12,
    delim_whitespace=True,
    engine='python',
)
uah_lo_df = pd.read_csv(data_url, **CSV_PARAMS)
```
