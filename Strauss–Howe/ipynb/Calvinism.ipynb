```python

def get_bio_row_dict(page_soup):
    row_dict = {}
    for param_th in page_soup.find_all('th', scope='row'):
        if (param_th.text.strip()=='Baptised') or (param_th.text.strip()=='Born') or (param_th.text.strip()=='Died'):
            row_dict['page_title'] = page_title
        if (param_th.text.strip()=='Baptised') or (param_th.text.strip()=='Born'):
            born_td = param_th.find_next_sibling('td')
            born_list = born_regex.findall(born_td.text)
            if len(born_list) > 0:
                born_x = max(born_list, key=lambda x: len(x[0]))[0]
                born_x = int(bc_regex.sub(r'-\1', born_x))
                row_dict['birth_year'] = born_x
        if (param_th.text.strip()=='Died'):
            died_td = param_th.find_next_sibling('td')
            died_list = born_regex.findall(died_td.text)
            if len(died_list) > 0:
                died_x = max(died_list, key=lambda x: len(x[0]))[0]
                died_x = int(bc_regex.sub(r'-\1', died_x))
                row_dict['death_year'] = died_x
    return row_dict
```

```python

def create_poi_df(rows_list):
    df = pd.DataFrame(rows_list)
    mask_series = df.duplicated(subset='page_title', keep='first')
    df = df[~mask_series].sort_values('birth_year').dropna(axis='index',
                                                            how='all').reset_index(drop=True)
    df['birth_diff'] = df.birth_year.diff()
    
    return df
```

```python

def enhance_poi_df(poi_df):
    generations_df = s.load_object('generations_df')
    f_str = '{}, {} {} {} of the {} generation in the {} saeculum.'
    for row_index, row_series in poi_df.iterrows():
        birth_year = row_series['birth_year']
        mask_series = (generations_df.birth_year_begin <= birth_year)
        mask_series = mask_series & (generations_df.birth_year_end >= birth_year)
        df = generations_df[mask_series]
        if df.shape[0] > 0:
            page_title = row_series.page_title
            generations_archetype = df.generations_archetype.tolist()[0]
            poi_df.loc[row_index, 'generations_archetype'] = generations_archetype
            turnings_archetype = df.turnings_archetype.tolist()[0]
            poi_df.loc[row_index, 'turnings_archetype'] = turnings_archetype
            generation_name = df.index.tolist()[0]
            poi_df.loc[row_index, 'generation_name'] = generation_name
            saeculum_name = df.saeculum_name.tolist()[0]
            poi_df.loc[row_index, 'saeculum_name'] = saeculum_name
            if generations_archetype[0] in ['A', 'I']:
                a_an = 'an'
            else:
                a_an = 'a'
            print(f_str.format(page_title, a_an, generations_archetype,
                               turnings_archetype, generation_name,
                               saeculum_name))

    return poi_df.sort_values('birth_year')
```

```python

def get_wikipedia_page_soup(page_title):
    page_soup = None
    try:
        page_obj = wikipedia.WikipediaPage(title=page_title)
        page_html = page_obj.html()
        page_soup = bs(page_html, 'html.parser')
    except Exception as e:
        print(e)
    
    return page_soup
```

```python

def get_toc_div(page_title):
    page_obj = wikipedia.WikipediaPage(title=page_title)
    page_html = page_obj.html()
    page_soup = bs(page_html, 'html.parser')
    toc_div = page_soup.find_all('div', id='toc')[0]
    
    return toc_div
```

```python

def get_poi_rows_list(toc_div):
    rows_list = []
    for sib_tag in toc_div.find_previous_siblings():
        #print(['sib_tag.{}'.format(fn) for fn in dir(sib_tag) if not fn.startswith('_')])
        if len(sib_tag.attrs) == 0:
            for a_tag in sib_tag.find_all(name='a', title=True):
                page_title = a_tag.attrs['title']
                page_soup = get_wikipedia_page_soup(page_title)
                row_dict = get_bio_row_dict(page_soup)
                rows_list.append(row_dict.copy())
    
    return rows_list
```

```python

def get_influences_list(page_title='Judith Butler'):
    influences_list = []
    try:
        page_obj = wikipedia.WikipediaPage(title=page_title)
        page_html = page_obj.html()
        page_soup = bs(page_html, 'html.parser')
        for NavHead_tag in page_soup.find_all(name='div', attrs={'class': 'NavHead'}):
            if NavHead_tag.text == 'Influences':
                for a_tag in NavHead_tag.parent.find_all(name='a', title=True):
                    influences_title = a_tag.attrs['title']
                    influences_list.append(influences_title)
        if len(influences_list) == 0:
            for param_th in page_soup.find_all('th', scope='row'):
                if (param_th.text.strip()=='Influences'):
                    influences_td = param_th.find_next_sibling('td')
                    for influences_link in influences_td.find_all('a', title=True):
                        influences_title = influences_link['title']
                        influences_list.append(influences_title)
    except Exception as e:
        print(e)
    
    return influences_list
```

```python

bs = wikipedia.BeautifulSoup
import networkx as nx

butler_digraph = nx.DiGraph()
node_set = set()
root_node = 'Judith Butler'
def build_graph(influences_digraph, page_title, n=0):
    node_set.add(page_title)
    if len(node_set) < 100:
        influences_list = get_influences_list(page_title=page_title)
        not_in_node_set_list = []
        for influencer_title in influences_list:
            if influencer_title not in node_set:
                not_in_node_set_list.append(influencer_title)
                influences_digraph.add_node(influencer_title)
                influences_digraph.add_edge(influencer_title, page_title)
        for influencer_title in not_in_node_set_list:
            node_set.add(influencer_title)
            if (n < 100) and (nx.shortest_path_length(butler_digraph, influencer_title, root_node) < 4):
                influences_digraph = build_graph(influences_digraph=influences_digraph,
                                                 page_title=influencer_title, n=n+1)
    
    return influences_digraph
```
