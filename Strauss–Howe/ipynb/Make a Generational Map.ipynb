```python

kml_str = '''<?xml version="1.0" encoding="UTF-8"?>
<kml xmlns="http://www.opengis.net/kml/2.2">
    <Document>
        <name>Babbitt Patriline</name>
        <open>1</open>
        <description>Lines generated by known birth and death locations</description>
        <Style id="thickBlackLine">
            <LineStyle>
                <color>87000000</color>
                <width>100</width>
            </LineStyle>
        </Style>
        <Style id="noDrivingDirections">
            <BalloonStyle>
                <text><![CDATA[
          <b>$[name]</b>
          <br /><br />
          $[description]
        ]]></text>
            </BalloonStyle>
        </Style>
        <Folder>
            <name>Paths</name>
            <visibility>1</visibility>
            <description>Generational paths.</description>{}
        </Folder>
    </Document>
</kml>'''
```

```python

placemark_str = '''
            <Placemark>
                <name>{}</name>
                <visibility>1</visibility>
                <styleUrl>#thickBlackLine</styleUrl>
                <LineString>
                    <extrude>1</extrude>
                    <tessellate>1</tessellate>
                    <altitudeMode>relativeToGround</altitudeMode>
                    <coordinates> {} </coordinates>
                </LineString>
            </Placemark>'''
```

```python

placemarks_str = ''
mask_series = patriline_df['Birth Latitude'].isnull() | patriline_df['Birth Longitude'].isnull()
mask_series = mask_series | patriline_df['Death Latitude'].isnull()
mask_series = mask_series | patriline_df['Death Longitude'].isnull()
for row_index, row_series in patriline_df[~mask_series].iterrows():
    coord_list = []
    for prefix in ['Birth', 'Death']:
        point_list = []
        for suffix in ['Latitude', 'Longitude']:
            column_name = '{} {}'.format(prefix, suffix)
            point_list.append(str(row_series[column_name]))
        point_list.append('6450')
        coord_list.append(','.join(point_list))
    placemarks_str += placemark_str.format(row_index, ' '.join(coord_list))
```

```python

kml_folder = os.path.join(SAVES_FOLDER, 'kml')
os.makedirs(name=kml_folder, exist_ok=True)
file_path = os.path.join(kml_folder, 'patriline.kml')
with open(file_path, 'w') as kmlfile:
    kmlfile.write(kml_str.format(placemarks_str))
print(os.path.abspath(file_path))
```
