```python

#page_title = 'List of United States presidential candidates by number of votes received'
#page_obj = wikipedia.page(page_title, redirect=False, preload=True)
#page_obj.html()
page_soup = su.get_page_soup(tables_url)
title_dict = {}
for link in page_soup.findAll('a', {'title': True}):
    try:
        title = link['href'].split('/')[-1].replace('_', ' ')
        text = link.string.strip()
        title_dict[text] = title
    except AttributeError as e:
        message = str(e).strip()
        print(f'AttributeError for {link}: {message}')
```

```python

from datetime import date

months_list = [date(2020, i, 1).strftime('%B') for i in range(1,13)]
'|'.join(months_list)
```

```python

candidates_list = candidate_year_df.sort_values('Candidate').Candidate.unique().tolist()
birth_years_dict = {}
for innerHTML in candidates_list:
    try:
        candidate = title_dict[innerHTML]
        summary_str = wikipedia.summary(candidate, auto_suggest=False)
        born_regex = re.compile(r'\bborn[^).]+(1(8|9)\d{2})')
        match_obj = born_regex.search(summary_str)
        if match_obj:
            born_year = int(match_obj.group(1))
        else:
            #born_regex = re.compile(r'\((' + '|'.join(months_list) + r') \d+, (\d+) â€“ (' + '|'.join(months_list) + r') \d+, \d+\)')
            born_regex = re.compile(r'(\d+) â€“ (' + '|'.join(months_list) + r')')
            match_obj = born_regex.search(summary_str)
            if match_obj:
                born_year = match_obj.group(1)
            else:
                born_year = np.nan
        print(f'{candidate}: {born_year}')
        birth_years_dict[innerHTML] = born_year
    except wikipedia.PageError as e:
        message = str(e).strip()
        print(f'PageError for {candidate}: {message}')
    except wikipedia.DisambiguationError as e:
        message = str(e).strip()
        print()
        print(f'DisambiguationError for {candidate}: {message}')
        print()
    except ValueError as e:
        message = str(e).strip()
        print(f'ValueError for {candidate}: {message}')
    except KeyError as e:
        message = str(e).strip()
        print(f'KeyError for {candidate}: {message}')
```

```python

candidate_year_df['Age'] = np.nan
for row_index, row_series in candidate_year_df.iterrows():
    candidate = row_series.Candidate
    if candidate in birth_years_dict:
        try:
            born_year = int(birth_years_dict[candidate])
            election_year = int(row_series.Year)
            candidate_year_df.loc[row_index, 'Age'] = election_year - born_year
        except ValueError as e:
            message = str(e).strip()
            print(f'ValueError for {candidate}: {message}')
            candidate_year_df.loc[row_index, 'Age'] = np.nan
```
