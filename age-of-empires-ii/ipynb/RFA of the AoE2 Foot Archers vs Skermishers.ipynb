```python

index_name = 'Unit Name'
dataframes_dict = load_dataframes(unit_availability_df='unit_availability_df')
unit_availability_df = dataframes_dict['unit_availability_df']
if 'Unnamed: 0' in unit_availability_df.columns:
    unit_availability_df = unit_availability_df.set_index('Unnamed: 0')
elif index_name in unit_availability_df.columns:
    unit_availability_df = unit_availability_df.set_index(index_name)
unit_availability_df.index.name = index_name
```

```python

column_descriptions_df = get_column_descriptions(unit_availability_df)
mask_series = column_descriptions_df['dtype'].isin(['int64', 'float64'])
numeric_columns_list = column_descriptions_df[mask_series]['column_name'].tolist()
print(column_descriptions_df.columns)
column_descriptions_df['dtype'].unique()
```

```python

# Basically, foot archers are in the archer armor class,
# not in the cavalry archer or cavalry armor class, and
# can be of the archer unit types but not the gunpowder unit types.
mask_series = True
mask_series = mask_series & (unit_availability_df['In_Archer_Armor_class'] == 1)
mask_series = mask_series & (unit_availability_df['In_Cavalry_archer_Armor_class'] == 0)
mask_series = mask_series & (unit_availability_df['In_Cavalry_Armor_class'] == 0)
mask_series = mask_series & (unit_availability_df['In_Gunpowder_unit_Unit_Type'] == 0)
foot_archers_list = unit_availability_df[mask_series].index.tolist()
columns_list = ['Heading', 'In_Gunpowder_unit_Unit_Type', 'In_Archer_Unit_Type', 'UpTo']
unit_availability_df[mask_series][columns_list].sort_index()
#sort_values(columns_list[1:], ascending=False)
```

```python

unit_availability_df[mask_series].sort_index().index.tolist()
```

```python

# JineappleAOE says:
# "Gunpowder units are not foot archers.
# The technical answer would be all units of the archer unit class.
# You could say that's all units affected by range upgrades that are on foot."
resource_set = set()
for column_name in unit_availability_df.columns:
    resource_boolean = ('archer' in column_name.lower())
    resource_boolean = resource_boolean or ('range' in column_name.lower())
    resource_boolean = resource_boolean or ('up' in column_name.lower())
    if resource_boolean:
        resource_set.add(column_name)
resource_set
```

```python

# Create the y variable
def is_foot_archer(row_series):
    if row_series.name in foot_archers_list:
        
        return True
    
    else:
        
        return False

y = unit_availability_df.apply(is_foot_archer, axis=1).values
```

```python

# Create the X variable
X = unit_availability_df[numeric_columns_list].fillna(0).values
```

```python

from sklearn.ensemble import RandomForestClassifier

regr = RandomForestClassifier(max_depth=5, n_estimators=100)
regr.fit(X, y)
top_6_tuple_list = sorted([(cn, ci) for cn, ci in zip(numeric_columns_list,
                                                      regr.feature_importances_) if ci > 0.0],
       key=lambda x: x[1], reverse=True)[:6]
```

```python

mask_series = (unit_availability_df.index.isin(foot_archers_list))
columns_list = [cn for (cn, cv) in top_6_tuple_list]
unit_availability_df[mask_series][columns_list].sort_values(columns_list, ascending=False)
```
